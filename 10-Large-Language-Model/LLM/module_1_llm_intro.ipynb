## 🚀 Chargement du modèle LLM (ex : mistralai/Mistral-7B-Instruct-v0.1)

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_id = "mistralai/Mistral-7B-Instruct-v0.1"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto")

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

## 💬 Exemple de génération de texte

```python
prompt = "Explique simplement ce qu'est un modèle de langage."

result = pipe(prompt, max_new_tokens=100, do_sample=True, top_k=50, temperature=0.7)
print(result[0]['generated_text'])
```
