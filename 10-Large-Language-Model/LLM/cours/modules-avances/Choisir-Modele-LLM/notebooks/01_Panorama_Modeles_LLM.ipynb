```python
# Comparaison interactive des mod√®les propri√©taires
proprietary_models = df_modeles[df_modeles['Type'] == 'Propri√©taire'].copy()

# Graphique radar des capacit√©s
fig = go.Figure()

categories = ['Performance (MMLU)', 'Vitesse', 'Rapport Qualit√©/Prix']

for _, model in proprietary_models.iterrows():
    # Normaliser les scores pour le radar
    performance = model['Score_MMLU'] / 100 * 100  # Score MMLU d√©j√† en %
    vitesse = model['Vitesse_Relative'] / 15 * 100  # Normaliser sur 15
    
    # Calculer rapport qualit√©/prix (inverse du co√ªt, normalis√©)
    if model['Cout_1M_Tokens'] > 0:
        qualite_prix = (1 / model['Cout_1M_Tokens']) * 100
    else:
        qualite_prix = 100  # Max pour les mod√®les gratuits
    
    values = [performance, vitesse, qualite_prix]
    
    fig.add_trace(go.Scatterpolar(
        r=values + [values[0]],  # Fermer le polygone
        theta=categories + [categories[0]],
        fill='toself',
        name=model['Mod√®le'],
        opacity=0.6
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 100]
        )
    ),
    title="üéØ Comparaison Radar des Mod√®les Propri√©taires",
    showlegend=True,
    height=500
)

fig.show()

print("üí° Plus l'aire est grande, mieux c'est !")
print("üìä Performance = Score MMLU, Vitesse = Rapidit√© relative, Qualit√©/Prix = Inverse du co√ªt")
```

## üìä HuggingFace Leaderboard - La R√©f√©rence

### üèÜ Qu'est-ce que le Leaderboard HuggingFace ?
Le **HuggingFace Open LLM Leaderboard** est la r√©f√©rence pour comparer les mod√®les open source.

### üìè M√©triques Principales :
1. **MMLU** (Massive Multitask Language Understanding)
   - üìö Questions acad√©miques multi-domaines
   - üéØ Mesure la connaissance g√©n√©rale

2. **HellaSwag** 
   - ü§î Compl√©tion de phrases avec bon sens
   - üí≠ Mesure le raisonnement de bon sens

3. **ARC** (AI2 Reasoning Challenge)
   - üßÆ Questions de science niveau √©cole
   - üî¨ Mesure le raisonnement scientifique

4. **TruthfulQA**
   - ‚úÖ Questions pi√®ge avec r√©ponses factuelles
   - üé≠ Mesure la v√©racit√© vs hallucinations

### üîç Comment Utiliser le Leaderboard :
1. **Allez sur** : https://huggingface.co/spaces/HuggingFaceH4/open_llm_leaderboard
2. **Filtrez** par taille de mod√®le qui vous int√©resse
3. **Regardez** le score global ET les m√©triques sp√©cifiques
4. **V√©rifiez** la date de soumission (plus r√©cent = mieux optimis√©)
5. **Lisez** les d√©tails du mod√®le (licence, usage commercial, etc.)

```python
# Simulation du leaderboard HuggingFace (donn√©es approximatives)
leaderboard_data = {
    'Mod√®le': [
        'Llama-2-70B-Chat', 'Mistral-7B-Instruct', 'Llama-2-13B-Chat', 
        'CodeLlama-34B-Instruct', 'Vicuna-13B', 'Llama-2-7B-Chat',
        'Alpaca-7B', 'Phi-3-Mini-4K', 'Gemma-7B', 'Mixtral-8x7B-Instruct'
    ],
    'Taille': ['70B', '7B', '13B', '34B', '13B', '7B', '7B', '3.8B', '7B', '47B'],
    'MMLU': [68.9, 62.5, 54.8, 53.7, 56.1, 45.3, 51.2, 68.8, 64.3, 70.6],
    'HellaSwag': [87.3, 83.3, 82.1, 76.2, 82.8, 78.6, 76.9, 82.4, 81.2, 87.6],
    'ARC': [64.6, 59.7, 59.4, 57.8, 59.1, 53.1, 52.3, 61.9, 61.1, 66.3],
    'TruthfulQA': [52.8, 50.3, 47.1, 43.6, 51.8, 38.2, 41.2, 44.1, 53.0, 54.1],
    'Score_Global': [68.4, 63.9, 60.9, 57.8, 62.5, 53.8, 55.4, 64.3, 64.9, 69.7]
}

df_leaderboard = pd.DataFrame(leaderboard_data)

# Heatmap du leaderboard
plt.figure(figsize=(12, 8))

# Pr√©parer les donn√©es pour la heatmap
metrics_cols = ['MMLU', 'HellaSwag', 'ARC', 'TruthfulQA', 'Score_Global']
heatmap_data = df_leaderboard[metrics_cols].values

# Cr√©er la heatmap
im = plt.imshow(heatmap_data, cmap='RdYlGn', aspect='auto')

# Configurer les axes
plt.xticks(range(len(metrics_cols)), metrics_cols)
plt.yticks(range(len(df_leaderboard)), df_leaderboard['Mod√®le'])
plt.title('üèÜ HuggingFace Leaderboard - Scores par M√©trique', fontsize=16, fontweight='bold', pad=20)

# Ajouter les valeurs dans les cellules
for i in range(len(df_leaderboard)):
    for j in range(len(metrics_cols)):
        text = plt.text(j, i, f'{heatmap_data[i, j]:.1f}',
                       ha="center", va="center", color="black", fontweight='bold')

# Colorbar
cbar = plt.colorbar(im)
cbar.set_label('Score (%)', rotation=270, labelpad=20)

plt.tight_layout()
plt.show()

# Top 5 par m√©trique
print("üèÖ TOP 3 par m√©trique :")
print("=" * 50)
for metric in metrics_cols:
    top3 = df_leaderboard.nlargest(3, metric)[['Mod√®le', 'Taille', metric]]
    print(f"\nüìä {metric}:")
    for i, (_, row) in enumerate(top3.iterrows(), 1):
        medal = ['ü•á', 'ü•à', 'ü•â'][i-1]
        print(f"  {medal} {row['Mod√®le']} ({row['Taille']}) - {row[metric]:.1f}%")
```
