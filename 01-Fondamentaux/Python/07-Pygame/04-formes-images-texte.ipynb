## 1. pygame.draw : Fonctions de Dessin

Le module `pygame.draw` contient toutes les fonctions pour dessiner des formes gÃ©omÃ©triques.

### Signature gÃ©nÃ©rale


- `surface` : sur quoi dessiner
- `couleur` : tuple RGB
- `width` : Ã©paisseur du trait (0 = rempli)

### 1.1 Rectangle : pygame.draw.rect()


### 1.6 Arc : pygame.draw.arc()

````python
%%writefile demo_formes.py
import pygame
import sys
import math

pygame.init()

LARGEUR, HAUTEUR = 1000, 700
ecran = pygame.display.set_mode((LARGEUR, HAUTEUR))
pygame.display.set_caption("DÃ©monstration des Formes")
clock = pygame.time.Clock()
font = pygame.font.Font(None, 24)

# Couleurs
NOIR = (0, 0, 0)
BLANC = (255, 255, 255)
ROUGE = (255, 0, 0)
VERT = (0, 255, 0)
BLEU = (0, 0, 255)
JAUNE = (255, 255, 0)
CYAN = (0, 255, 255)
MAGENTA = (255, 0, 255)
ORANGE = (255, 165, 0)

running = True
while running:
    clock.tick(60)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Dessiner
    ecran.fill(NOIR)
    
    # Rectangle rempli
    pygame.draw.rect(ecran, ROUGE, (50, 50, 150, 100))
    texte = font.render("Rectangle rempli", True, BLANC)
    ecran.blit(texte, (50, 160))
    
    # Rectangle avec contour
    pygame.draw.rect(ecran, VERT, (250, 50, 150, 100), width=3)
    texte = font.render("Rectangle contour", True, BLANC)
    ecran.blit(texte, (250, 160))
    
    # Rectangle coins arrondis
    pygame.draw.rect(ecran, BLEU, (450, 50, 150, 100), border_radius=15)
    texte = font.render("Coins arrondis", True, BLANC)
    ecran.blit(texte, (450, 160))
    
    # Cercle rempli
    pygame.draw.circle(ecran, JAUNE, (725, 100), 50)
    texte = font.render("Cercle rempli", True, BLANC)
    ecran.blit(texte, (675, 160))
    
    # Cercle contour
    pygame.draw.circle(ecran, CYAN, (900, 100), 50, width=4)
    texte = font.render("Cercle contour", True, BLANC)
    ecran.blit(texte, (850, 160))
    
    # Lignes
    pygame.draw.line(ecran, MAGENTA, (50, 250), (200, 250), width=1)
    pygame.draw.line(ecran, MAGENTA, (50, 270), (200, 270), width=3)
    pygame.draw.line(ecran, MAGENTA, (50, 295), (200, 295), width=7)
    texte = font.render("Lignes", True, BLANC)
    ecran.blit(texte, (50, 310))
    
    # Triangle (polygone)
    points_triangle = [(350, 250), (300, 340), (400, 340)]
    pygame.draw.polygon(ecran, ORANGE, points_triangle)
    texte = font.render("Triangle", True, BLANC)
    ecran.blit(texte, (320, 350))
    
    # Pentagone
    centre_x, centre_y = 550, 290
    rayon = 60
    points_penta = []
    for i in range(5):
        angle = i * 2 * math.pi / 5 - math.pi / 2
        x = centre_x + rayon * math.cos(angle)
        y = centre_y + rayon * math.sin(angle)
        points_penta.append((x, y))
    pygame.draw.polygon(ecran, VERT, points_penta, width=3)
    texte = font.render("Pentagone", True, BLANC)
    ecran.blit(texte, (510, 360))
    
    # Ellipse
    pygame.draw.ellipse(ecran, BLEU, (700, 250, 150, 80))
    texte = font.render("Ellipse", True, BLANC)
    ecran.blit(texte, (750, 340))
    
    # Arc
    pygame.draw.arc(ecran, ROUGE, (50, 400, 150, 150), 0, math.pi, width=4)
    texte = font.render("Arc", True, BLANC)
    ecran.blit(texte, (100, 560))
    
    # Ã‰toile (lignes multiples)
    centre_x, centre_y = 350, 475
    rayon_ext = 70
    rayon_int = 30
    points_etoile = []
    for i in range(10):
        angle = i * math.pi / 5 - math.pi / 2
        rayon = rayon_ext if i % 2 == 0 else rayon_int
        x = centre_x + rayon * math.cos(angle)
        y = centre_y + rayon * math.sin(angle)
        points_etoile.append((x, y))
    pygame.draw.polygon(ecran, JAUNE, points_etoile)
    texte = font.render("Ã‰toile", True, BLANC)
    ecran.blit(texte, (320, 560))
    
    # Multiple lignes (polyline)
    points_ligne = [(550, 450), (600, 400), (650, 450), (700, 420), (750, 480)]
    pygame.draw.lines(ecran, CYAN, False, points_ligne, width=3)
    texte = font.render("Polyline", True, BLANC)
    ecran.blit(texte, (620, 490))
    
    pygame.display.flip()

pygame.quit()
sys.exit()
````

### Pourquoi convert() ?

- `convert()` : convertit l'image au format de l'Ã©cran â†’ **beaucoup plus rapide**
- `convert_alpha()` : pareil mais garde la transparence (canal alpha)
- **Toujours** utiliser convert() aprÃ¨s le chargement!

````python
%%writefile demo_images_procedurales.py
import pygame
import sys

pygame.init()

LARGEUR, HAUTEUR = 800, 600
ecran = pygame.display.set_mode((LARGEUR, HAUTEUR))
pygame.display.set_caption("Images ProcÃ©durales")
clock = pygame.time.Clock()

# CrÃ©er une image procÃ©durale (damier)
def creer_damier(taille, taille_case, couleur1, couleur2):
    surface = pygame.Surface((taille, taille))
    nb_cases = taille // taille_case
    
    for i in range(nb_cases):
        for j in range(nb_cases):
            couleur = couleur1 if (i + j) % 2 == 0 else couleur2
            rect = (i * taille_case, j * taille_case, taille_case, taille_case)
            pygame.draw.rect(surface, couleur, rect)
    
    return surface.convert()

# CrÃ©er un cercle colorÃ©
def creer_cercle(rayon, couleur):
    surface = pygame.Surface((rayon * 2, rayon * 2))
    surface.fill((0, 0, 0))
    surface.set_colorkey((0, 0, 0))  # Noir transparent
    pygame.draw.circle(surface, couleur, (rayon, rayon), rayon)
    return surface.convert()

# CrÃ©er les images
damier = creer_damier(200, 25, (255, 255, 255), (200, 200, 200))
cercle_rouge = creer_cercle(50, (255, 0, 0))
cercle_vert = creer_cercle(40, (0, 255, 0))

# Position
x, y = 400, 300

running = True
while running:
    dt = clock.tick(60) / 1000.0
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Mouvement
    keys = pygame.key.get_pressed()
    vitesse = 200
    if keys[pygame.K_LEFT]: x -= vitesse * dt
    if keys[pygame.K_RIGHT]: x += vitesse * dt
    if keys[pygame.K_UP]: y -= vitesse * dt
    if keys[pygame.K_DOWN]: y += vitesse * dt
    
    # Dessiner
    ecran.fill((50, 50, 100))
    
    # Damier en fond
    ecran.blit(damier, (50, 50))
    ecran.blit(damier, (550, 50))
    ecran.blit(damier, (50, 350))
    ecran.blit(damier, (550, 350))
    
    # Cercles qui suivent la souris
    mouse_x, mouse_y = pygame.mouse.get_pos()
    ecran.blit(cercle_rouge, (mouse_x - 50, mouse_y - 50))
    
    # Cercle contrÃ´lÃ© au clavier
    ecran.blit(cercle_vert, (x - 40, y - 40))
    
    pygame.display.flip()

pygame.quit()
sys.exit()
````

### 3.4 Rotation optimisÃ©e : rotozoom()

```python

```

````python
%%writefile demo_transformations.py
import pygame
import sys
import math

pygame.init()

LARGEUR, HAUTEUR = 1000, 700
ecran = pygame.display.set_mode((LARGEUR, HAUTEUR))
pygame.display.set_caption("Transformations d'Images")
clock = pygame.time.Clock()
font = pygame.font.Font(None, 24)

# CrÃ©er une image de base (flÃ¨che)
def creer_fleche(taille):
    surface = pygame.Surface((taille, taille))
    surface.fill((255, 0, 255))  # Magenta = transparent
    surface.set_colorkey((255, 0, 255))
    
    # Dessiner une flÃ¨che pointant vers le haut
    centre = taille // 2
    points = [
        (centre, 10),
        (taille - 10, taille - 10),
        (centre + 10, taille - 20),
        (centre + 10, taille // 2),
        (centre - 10, taille // 2),
        (centre - 10, taille - 20),
        (10, taille - 10)
    ]
    pygame.draw.polygon(surface, (0, 150, 255), points)
    return surface.convert()

# Image originale
image_originale = creer_fleche(100)

# Angle de rotation
angle = 0

running = True
while running:
    dt = clock.tick(60) / 1000.0
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # IncrÃ©menter l'angle
    angle += 30 * dt  # 30 degrÃ©s par seconde
    
    # Dessiner
    ecran.fill((30, 30, 50))
    
    # Originale
    ecran.blit(image_originale, (50, 50))
    texte = font.render("Original", True, (255, 255, 255))
    ecran.blit(texte, (60, 160))
    
    # Scale x2
    image_2x = pygame.transform.scale(image_originale, (200, 200))
    ecran.blit(image_2x, (200, 50))
    texte = font.render("Scale x2", True, (255, 255, 255))
    ecran.blit(texte, (260, 260))
    
    # Scale x0.5
    image_05x = pygame.transform.scale(image_originale, (50, 50))
    ecran.blit(image_05x, (450, 75))
    texte = font.render("Scale x0.5", True, (255, 255, 255))
    ecran.blit(texte, (450, 135))
    
    # Rotation
    image_tournee = pygame.transform.rotate(image_originale, angle)
    rect = image_tournee.get_rect(center=(650, 100))
    ecran.blit(image_tournee, rect)
    texte = font.render(f"Rotate {int(angle) % 360}Â°", True, (255, 255, 255))
    ecran.blit(texte, (600, 180))
    
    # Rotozoom (rotation + zoom)
    echelle = 1.5 + 0.5 * math.sin(angle * 0.05)
    image_rotozoom = pygame.transform.rotozoom(image_originale, -angle * 2, echelle)
    rect = image_rotozoom.get_rect(center=(850, 100))
    ecran.blit(image_rotozoom, rect)
    texte = font.render("Rotozoom", True, (255, 255, 255))
    ecran.blit(texte, (810, 180))
    
    # Flip horizontal
    image_flip_h = pygame.transform.flip(image_originale, True, False)
    ecran.blit(image_flip_h, (50, 300))
    texte = font.render("Flip H", True, (255, 255, 255))
    ecran.blit(texte, (60, 410))
    
    # Flip vertical
    image_flip_v = pygame.transform.flip(image_originale, False, True)
    ecran.blit(image_flip_v, (200, 300))
    texte = font.render("Flip V", True, (255, 255, 255))
    ecran.blit(texte, (210, 410))
    
    # Flip H+V
    image_flip_hv = pygame.transform.flip(image_originale, True, True)
    ecran.blit(image_flip_hv, (350, 300))
    texte = font.render("Flip H+V", True, (255, 255, 255))
    ecran.blit(texte, (350, 410))
    
    # Combinaison : flip + rotate
    image_combo = pygame.transform.flip(image_originale, True, False)
    image_combo = pygame.transform.rotate(image_combo, 45)
    rect = image_combo.get_rect(center=(550, 350))
    ecran.blit(image_combo, rect)
    texte = font.render("Flip+Rotate", True, (255, 255, 255))
    ecran.blit(texte, (500, 430))
    
    # Scale dÃ©formÃ© (non proportionnel)
    image_deforme = pygame.transform.scale(image_originale, (150, 50))
    ecran.blit(image_deforme, (700, 325))
    texte = font.render("Scale dÃ©formÃ©", True, (255, 255, 255))
    ecran.blit(texte, (710, 385))
    
    pygame.display.flip()

pygame.quit()
sys.exit()
````

````python
%%writefile demo_texte.py
import pygame
import sys

pygame.init()

LARGEUR, HAUTEUR = 800, 700
ecran = pygame.display.set_mode((LARGEUR, HAUTEUR))
pygame.display.set_caption("DÃ©monstration de Texte")
clock = pygame.time.Clock()

# DiffÃ©rentes polices et tailles
font_small = pygame.font.Font(None, 24)
font_medium = pygame.font.Font(None, 48)
font_large = pygame.font.Font(None, 72)

# Police systÃ¨me
font_arial = pygame.font.SysFont('arial', 36)
font_courier = pygame.font.SysFont('courier', 32, bold=True)

# Couleurs
NOIR = (0, 0, 0)
BLANC = (255, 255, 255)
ROUGE = (255, 0, 0)
VERT = (0, 255, 0)
BLEU = (0, 0, 255)
JAUNE = (255, 255, 0)

# Compteur pour texte dynamique
compteur = 0

running = True
while running:
    dt = clock.tick(60) / 1000.0
    compteur += dt
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Dessiner
    ecran.fill(NOIR)
    
    # Titre centrÃ©
    titre = font_large.render("PYGAME TEXT", True, JAUNE)
    rect = titre.get_rect(center=(LARGEUR//2, 50))
    ecran.blit(titre, rect)
    
    # DiffÃ©rentes tailles
    y = 130
    for taille, font in [(24, font_small), (48, font_medium), (72, font_large)]:
        texte = font.render(f"Taille {taille}", True, BLANC)
        ecran.blit(texte, (50, y))
        y += taille + 10
    
    # DiffÃ©rentes couleurs
    y = 130
    for nom, couleur in [("Rouge", ROUGE), ("Vert", VERT), ("Bleu", BLEU)]:
        texte = font_medium.render(nom, True, couleur)
        ecran.blit(texte, (450, y))
        y += 60
    
    # Avec fond
    texte_fond = font_arial.render("Texte avec fond", True, NOIR, JAUNE)
    ecran.blit(texte_fond, (50, 400))
    
    # Sans antialiasing
    texte_aa = font_medium.render("Avec AA", True, BLANC)
    texte_no_aa = font_medium.render("Sans AA", False, BLANC)
    ecran.blit(texte_aa, (50, 460))
    ecran.blit(texte_no_aa, (50, 520))
    
    # Texte dynamique
    fps_text = font_small.render(f"FPS: {clock.get_fps():.1f}", True, VERT)
    ecran.blit(fps_text, (10, 10))
    
    temps_text = font_small.render(f"Temps: {int(compteur)}s", True, VERT)
    ecran.blit(temps_text, (10, 35))
    
    # Position souris
    mouse_x, mouse_y = pygame.mouse.get_pos()
    mouse_text = font_small.render(f"Souris: ({mouse_x}, {mouse_y})", True, BLANC)
    rect = mouse_text.get_rect(bottomright=(LARGEUR-10, HAUTEUR-10))
    ecran.blit(mouse_text, rect)
    
    # Texte qui suit la souris
    follow_text = font_courier.render("Je te suis!", True, ROUGE)
    rect = follow_text.get_rect(center=(mouse_x, mouse_y - 30))
    ecran.blit(follow_text, rect)
    
    pygame.display.flip()

pygame.quit()
sys.exit()
````

````python
%%writefile solution_ex2_texte.py
import pygame
import sys

pygame.init()

LARGEUR, HAUTEUR = 800, 600
ecran = pygame.display.set_mode((LARGEUR, HAUTEUR))
pygame.display.set_caption("Exercice 2 : Texte")
clock = pygame.time.Clock()

font_titre = pygame.font.Font(None, 72)
font_normal = pygame.font.Font(None, 36)

NOIR = (0, 0, 0)
couleurs = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (255, 255, 0)]
compteur = 0
index_couleur = 0
dernier_changement = 0

running = True
while running:
    dt = clock.tick(60) / 1000.0
    compteur += dt
    
    # Changer couleur toutes les secondes
    if compteur - dernier_changement >= 1.0:
        index_couleur = (index_couleur + 1) % len(couleurs)
        dernier_changement = compteur
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    ecran.fill(NOIR)
    
    # Titre centrÃ©
    titre = font_titre.render("MON JEU", True, couleurs[index_couleur])
    rect = titre.get_rect(center=(LARGEUR//2, 50))
    ecran.blit(titre, rect)
    
    # FPS
    fps_text = font_normal.render(f"FPS: {clock.get_fps():.0f}", True, (0, 255, 0))
    ecran.blit(fps_text, (10, 10))
    
    # Compteur
    count_text = font_normal.render(f"Temps: {int(compteur)}s", True, (255, 255, 255))
    rect = count_text.get_rect(center=(LARGEUR//2, HAUTEUR//2))
    ecran.blit(count_text, rect)
    
    # Instructions
    instr = font_normal.render("ESC pour quitter", True, (200, 200, 200))
    rect = instr.get_rect(center=(LARGEUR//2, HAUTEUR-30))
    ecran.blit(instr, rect)
    
    pygame.display.flip()

pygame.quit()
sys.exit()
````

````python
%%writefile solution_ex3_rotation.py
import pygame
import sys
import math

pygame.init()

LARGEUR, HAUTEUR = 800, 600
ecran = pygame.display.set_mode((LARGEUR, HAUTEUR))
pygame.display.set_caption("Exercice 3 : Rotation")
clock = pygame.time.Clock()
font = pygame.font.Font(None, 36)

# CrÃ©er une Ã©toile
def creer_etoile(taille):
    surface = pygame.Surface((taille, taille))
    surface.fill((0, 0, 0))
    surface.set_colorkey((0, 0, 0))
    
    centre = taille // 2
    rayon_ext = taille // 2 - 10
    rayon_int = rayon_ext // 2
    points = []
    
    for i in range(10):
        angle = i * math.pi / 5 - math.pi / 2
        rayon = rayon_ext if i % 2 == 0 else rayon_int
        x = centre + rayon * math.cos(angle)
        y = centre + rayon * math.sin(angle)
        points.append((x, y))
    
    pygame.draw.polygon(surface, (255, 215, 0), points)
    return surface.convert()

image_originale = creer_etoile(150)
angle = 0
vitesse = 60  # degrÃ©s par seconde
direction = 1

running = True
while running:
    dt = clock.tick(60) / 1000.0
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                direction *= -1
    
    # Vitesse
    keys = pygame.key.get_pressed()
    if keys[pygame.K_UP]:
        vitesse = min(vitesse + 100 * dt, 300)
    if keys[pygame.K_DOWN]:
        vitesse = max(vitesse - 100 * dt, 0)
    
    # Rotation
    angle += vitesse * direction * dt
    
    # Dessiner
    ecran.fill((30, 30, 60))
    
    # Image tournÃ©e
    image_tournee = pygame.transform.rotate(image_originale, angle)
    rect = image_tournee.get_rect(center=(LARGEUR//2, HAUTEUR//2))
    ecran.blit(image_tournee, rect)
    
    # Info
    angle_text = font.render(f"Angle: {int(angle) % 360}Â°", True, (255, 255, 255))
    vitesse_text = font.render(f"Vitesse: {int(vitesse)}Â°/s", True, (255, 255, 255))
    dir_text = font.render(f"Direction: {'Horaire' if direction == -1 else 'Anti-horaire'}", True, (255, 255, 255))
    
    ecran.blit(angle_text, (10, 10))
    ecran.blit(vitesse_text, (10, 50))
    ecran.blit(dir_text, (10, 90))
    
    pygame.display.flip()

pygame.quit()
sys.exit()
````
