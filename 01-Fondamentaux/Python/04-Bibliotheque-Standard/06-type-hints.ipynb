## 1. Pourquoi les type hints ?

Les type hints (annotations de types) ne sont PAS du typage statique ! Python reste dynamiquement typÃ©.

### Avantages des type hints :

1. **Documentation** : Le code s'auto-documente
2. **IDE** : AutocomplÃ©tion et dÃ©tection d'erreurs
3. **Outils statiques** : mypy, pyright, pytype
4. **Refactoring** : Plus sÃ»r et plus facile
5. **Maintenance** : Code plus lisible pour les Ã©quipes

```python
# Sans type hints - difficile Ã  comprendre
def process_data(data, config):
    return data * config['multiplier'] + config['offset']

# Avec type hints - intentions claires
def process_data_typed(data: float, config: dict[str, float]) -> float:
    return data * config['multiplier'] + config['offset']

# Test
config = {'multiplier': 2.5, 'offset': 10.0}
result = process_data_typed(5.0, config)
print(f"RÃ©sultat: {result}")
```

## 2. Annotations de base

Python propose des types de base pour les annotations.

```python
# Types primitifs
age: int = 30
nom: str = "Alice"
taille: float = 1.75
actif: bool = True
rien: None = None

print(f"Age: {age} (type: {type(age).__name__})")
print(f"Nom: {nom} (type: {type(nom).__name__})")
print(f"Taille: {taille} (type: {type(taille).__name__})")
print(f"Actif: {actif} (type: {type(actif).__name__})")
print(f"Rien: {rien} (type: {type(rien).__name__})")

# IMPORTANT: Les annotations ne forcent PAS le type!
age: int = "trente"  # Pas d'erreur Ã  l'exÃ©cution!
print(f"\nAge (incorrect): {age} (type: {type(age).__name__})")
```

## 3. Annotations de fonctions

Syntaxe : `def fonction(param: Type) -> ReturnType:`

```python
# Fonction simple
def saluer(nom: str) -> str:
    return f"Bonjour {nom}!"

# Plusieurs paramÃ¨tres
def additionner(a: int, b: int) -> int:
    return a + b

# ParamÃ¨tres avec valeurs par dÃ©faut
def diviser(a: float, b: float, arrondir: bool = False) -> float:
    resultat = a / b
    return round(resultat, 2) if arrondir else resultat

# Fonction sans retour
def afficher(message: str) -> None:
    print(message)

# Tests
print(saluer("Alice"))
print(f"Addition: {additionner(5, 3)}")
print(f"Division: {diviser(10, 3)}")
print(f"Division arrondie: {diviser(10, 3, arrondir=True)}")
afficher("Message affichÃ©")
```

## 5. GÃ©nÃ©riques modernes (Python 3.9+)

Python 3.9+ permet d'utiliser directement `list`, `dict`, `tuple`, `set`.

```python
# Python 3.9+ - syntaxe moderne (recommandÃ©e)
def traiter_nombres(nombres: list[int]) -> dict[str, int]:
    return {
        'somme': sum(nombres),
        'max': max(nombres),
        'min': min(nombres),
        'count': len(nombres)
    }

# Dict avec types de clÃ© et valeur
def compter_mots(texte: str) -> dict[str, int]:
    mots = texte.lower().split()
    compteur: dict[str, int] = {}
    for mot in mots:
        compteur[mot] = compteur.get(mot, 0) + 1
    return compteur

# Liste de tuples
def obtenir_villes() -> list[tuple[str, int]]:
    return [
        ("Paris", 2_165_423),
        ("Marseille", 869_815),
        ("Lyon", 513_275)
    ]

# Nested types
def obtenir_config() -> dict[str, list[str] | int]:
    return {
        'hosts': ['localhost', '127.0.0.1'],
        'port': 8080,
        'allowed_ips': ['192.168.1.1', '192.168.1.2']
    }

# Tests
print(f"Traitement: {traiter_nombres([10, 20, 30])}")
print(f"\nComptage: {compter_mots('python est gÃ©nial python est puissant')}")
print(f"\nVilles: {obtenir_villes()}")
print(f"\nConfig: {obtenir_config()}")
```

## 6. Optional et Union

- `Optional[X]` = `X | None` (valeur peut Ãªtre None)
- `Union[X, Y]` = `X | Y` (valeur peut Ãªtre X ou Y)

```python
from typing import Optional

# Optional - Python < 3.10
def rechercher_utilisateur(id: int) -> Optional[str]:
    users = {1: "Alice", 2: "Bob", 3: "Charlie"}
    return users.get(id)  # Retourne None si non trouvÃ©

# Optional - Python 3.10+ (syntaxe moderne)
def rechercher_utilisateur_v2(id: int) -> str | None:
    users = {1: "Alice", 2: "Bob", 3: "Charlie"}
    return users.get(id)

# Union - Python < 3.10
from typing import Union

def formater_valeur(valeur: Union[int, float, str]) -> str:
    if isinstance(valeur, (int, float)):
        return f"Nombre: {valeur}"
    return f"Texte: {valeur}"

# Union - Python 3.10+ (syntaxe moderne)
def formater_valeur_v2(valeur: int | float | str) -> str:
    if isinstance(valeur, (int, float)):
        return f"Nombre: {valeur}"
    return f"Texte: {valeur}"

# Tests
print(f"User 1: {rechercher_utilisateur(1)}")
print(f"User 99: {rechercher_utilisateur(99)}")
print(f"\n{formater_valeur(42)}")
print(f"{formater_valeur(3.14)}")
print(f"{formater_valeur('hello')}")
```

## 7. TypeVar et gÃ©nÃ©riques custom

Pour crÃ©er des fonctions gÃ©nÃ©riques qui fonctionnent avec n'importe quel type.

```python
from typing import TypeVar

# TypeVar pour crÃ©er un type gÃ©nÃ©rique
T = TypeVar('T')

def premier_element(liste: list[T]) -> T | None:
    """Retourne le premier Ã©lÃ©ment d'une liste, ou None si vide."""
    return liste[0] if liste else None

def dernier_element(liste: list[T]) -> T | None:
    """Retourne le dernier Ã©lÃ©ment d'une liste, ou None si vide."""
    return liste[-1] if liste else None

# TypeVar avec contraintes
Nombre = TypeVar('Nombre', int, float)

def doubler(valeur: Nombre) -> Nombre:
    """Double un nombre (int ou float)."""
    return valeur * 2

# Tests
nombres = [1, 2, 3, 4, 5]
textes = ["a", "b", "c"]

print(f"Premier nombre: {premier_element(nombres)}")
print(f"Premier texte: {premier_element(textes)}")
print(f"Dernier nombre: {dernier_element(nombres)}")
print(f"Liste vide: {premier_element([])}")

print(f"\nDoubler int: {doubler(5)}")
print(f"Doubler float: {doubler(3.14)}")
```

## 8. Literal, TypeAlias, Final

Types avancÃ©s pour des cas spÃ©cifiques.

```python
from typing import Literal, Final

# Literal - valeurs spÃ©cifiques uniquement
def traiter_mode(mode: Literal["read", "write", "append"]) -> str:
    return f"Mode: {mode}"

# TypeAlias - crÃ©er des alias de types (Python 3.10+)
# type UserId = int  # Python 3.12+
UserId = int  # Python < 3.12
Coordonnees = tuple[float, float]

def obtenir_position(user_id: UserId) -> Coordonnees:
    positions = {1: (48.8566, 2.3522), 2: (43.2965, 5.3698)}
    return positions.get(user_id, (0.0, 0.0))

# Final - constante qui ne devrait pas Ãªtre modifiÃ©e
MAX_CONNECTIONS: Final[int] = 100
API_VERSION: Final[str] = "v1.0"

# Tests
print(traiter_mode("read"))
print(traiter_mode("write"))
# print(traiter_mode("delete"))  # Erreur avec mypy!

print(f"\nPosition user 1: {obtenir_position(1)}")
print(f"Max connections: {MAX_CONNECTIONS}")
print(f"API version: {API_VERSION}")
```

## 9. Type hints vs Runtime

**CRITIQUE** : Les type hints ne sont PAS vÃ©rifiÃ©s Ã  l'exÃ©cution !

```python
# Cette fonction s'exÃ©cute SANS erreur!
def additionner(a: int, b: int) -> int:
    return a + b

# Appels incorrects - AUCUNE erreur Ã  l'exÃ©cution!
result1 = additionner("hello", "world")  # Types incorrects
print(f"RÃ©sultat 1: {result1}")  # helloworld

result2 = additionner(5, 3.14)  # int + float
print(f"RÃ©sultat 2: {result2}")  # 8.14

result3 = additionner([1, 2], [3, 4])  # lists
print(f"RÃ©sultat 3: {result3}")  # [1, 2, 3, 4]

# Les type hints sont dans __annotations__
print(f"\nAnnotations: {additionner.__annotations__}")

# Pour vÃ©rifier Ã  l'exÃ©cution (si vraiment nÃ©cessaire)
def additionner_strict(a: int, b: int) -> int:
    if not isinstance(a, int) or not isinstance(b, int):
        raise TypeError(f"Attendu int, reÃ§u {type(a).__name__} et {type(b).__name__}")
    return a + b

try:
    additionner_strict("hello", "world")
except TypeError as e:
    print(f"\nErreur: {e}")
```

### 2. Optional != paramÃ¨tre optionnel

```python
from typing import Optional

# PIÃˆGE: Optional ne rend PAS le paramÃ¨tre optionnel!
def afficher_message(message: Optional[str]) -> None:
    if message is not None:
        print(message)

# Doit quand mÃªme passer un argument
afficher_message(None)  # OK
afficher_message("Hello")  # OK
# afficher_message()  # ERREUR: argument manquant!

# SOLUTION: Combiner avec valeur par dÃ©faut
def afficher_message_v2(message: Optional[str] = None) -> None:
    if message is not None:
        print(message)

afficher_message_v2()  # OK maintenant
afficher_message_v2("Hello")
```

### 3. Mutable default values

```python
# PIÃˆGE: Valeur par dÃ©faut mutable
def ajouter_element_mauvais(element: str, liste: list[str] = []) -> list[str]:
    liste.append(element)
    return liste

# La liste est partagÃ©e entre les appels!
result1 = ajouter_element_mauvais("a")
result2 = ajouter_element_mauvais("b")
print(f"Result1: {result1}")  # ['a', 'b'] - INATTENDU!
print(f"Result2: {result2}")  # ['a', 'b']
print(f"MÃªme objet? {result1 is result2}")  # True

# SOLUTION: Utiliser None comme dÃ©faut
def ajouter_element_bon(element: str, liste: list[str] | None = None) -> list[str]:
    if liste is None:
        liste = []
    liste.append(element)
    return liste

result3 = ajouter_element_bon("a")
result4 = ajouter_element_bon("b")
print(f"\nResult3: {result3}")  # ['a'] - OK!
print(f"Result4: {result4}")  # ['b'] - OK!
print(f"MÃªme objet? {result3 is result4}")  # False
```

### 4. Any vs object

```python
from typing import Any

# Any - dÃ©sactive la vÃ©rification de type
def traiter_any(valeur: Any) -> Any:
    return valeur.upper()  # mypy n'Ã©met pas d'avertissement

# object - type le plus gÃ©nÃ©ral, mais vÃ©rifiÃ©
def traiter_object(valeur: object) -> str:
    # return valeur.upper()  # mypy Ã©met une erreur!
    if isinstance(valeur, str):
        return valeur.upper()
    return str(valeur)

# RECOMMANDATION: Ã‰viter Any, utiliser Union ou TypeVar
def traiter_valeur(valeur: str | int | float) -> str:
    if isinstance(valeur, str):
        return valeur.upper()
    return str(valeur)

print(traiter_valeur("hello"))
print(traiter_valeur(42))
print(traiter_valeur(3.14))
```

### Exercice 3: SystÃ¨me de cache typÃ©

CrÃ©ez un systÃ¨me de cache simple avec type hints appropriÃ©s.

```python
# Ã€ complÃ©ter avec type hints
class Cache:
    def __init__(self):
        self._cache = {}
    
    def set(self, key, value):
        self._cache[key] = value
    
    def get(self, key, default=None):
        return self._cache.get(key, default)
    
    def has(self, key):
        return key in self._cache
    
    def clear(self):
        self._cache.clear()

# Test
cache = Cache()
cache.set("user_1", {"name": "Alice", "age": 30})
print(cache.get("user_1"))
print(cache.has("user_1"))
print(cache.get("user_2", "Not found"))
```

### Solution Exercice 3

```python
from typing import TypeVar, Generic, Optional

K = TypeVar('K')  # Type de la clÃ©
V = TypeVar('V')  # Type de la valeur

class Cache(Generic[K, V]):
    """Cache gÃ©nÃ©rique avec types paramÃ©trÃ©s."""
    
    def __init__(self) -> None:
        self._cache: dict[K, V] = {}
    
    def set(self, key: K, value: V) -> None:
        """Ajoute une valeur au cache."""
        self._cache[key] = value
    
    def get(self, key: K, default: V | None = None) -> V | None:
        """RÃ©cupÃ¨re une valeur du cache."""
        return self._cache.get(key, default)
    
    def has(self, key: K) -> bool:
        """VÃ©rifie si une clÃ© existe dans le cache."""
        return key in self._cache
    
    def clear(self) -> None:
        """Vide le cache."""
        self._cache.clear()
    
    def size(self) -> int:
        """Retourne la taille du cache."""
        return len(self._cache)

# Tests avec types spÃ©cifiques
# Cache pour users (str -> dict)
user_cache: Cache[str, dict[str, str | int]] = Cache()
user_cache.set("user_1", {"name": "Alice", "age": 30})
user_cache.set("user_2", {"name": "Bob", "age": 25})

print(f"User 1: {user_cache.get('user_1')}")
print(f"Has user_2: {user_cache.has('user_2')}")
print(f"User 3 (default): {user_cache.get('user_3', 'Not found')}")
print(f"Taille du cache: {user_cache.size()}")

# Cache pour nombres (int -> float)
number_cache: Cache[int, float] = Cache()
number_cache.set(1, 3.14)
number_cache.set(2, 2.71)

print(f"\nNombre 1: {number_cache.get(1)}")
print(f"Nombre 3: {number_cache.get(3, 0.0)}")
```
