## 6. uv : le gestionnaire moderne

uv est un gestionnaire de packages Python ultra-rapide Ã©crit en Rust.

````python
print("UV : LE GESTIONNAIRE MODERNE")
print("=" * 60)

print("""
uv est un remplaÃ§ant de pip/pip-tools, Ã©crit en Rust.

AVANTAGES:
  âœ… 10-100x plus rapide que pip
  âœ… RÃ©solution de dÃ©pendances amÃ©liorÃ©e
  âœ… Cache global intelligent
  âœ… Compatible avec pip (drop-in replacement)
  âœ… Support de pyproject.toml
  âœ… Lock file automatique

INSTALLATION:
  # Linux/macOS:
  curl -LsSf https://astral.sh/uv/install.sh | sh
  
  # Windows:
  powershell -c "irm https://astral.sh/uv/install.ps1 | iex"
  
  # Avec pip:
  pip install uv
""")

print("\nCOMMANDES UV:")
print("=" * 60)

commandes_uv = """
# Environnement virtuel
uv venv                        # CrÃ©er venv
uv venv --python 3.11          # Avec version spÃ©cifique

# Installation
uv pip install requests        # Installer un package
uv pip install -r requirements.txt
uv pip install -e .            # Mode Ã©ditable

# Mise Ã  jour
uv pip install --upgrade requests

# Compilation de dÃ©pendances
uv pip compile requirements.in -o requirements.txt

# Synchronisation (comme pip-sync)
uv pip sync requirements.txt

# Liste
uv pip list
uv pip freeze
"""

print(commandes_uv)

print("\nCOMPARAISON DES PERFORMANCES:")
print("  pip install numpy pandas: ~45s")
print("  uv pip install numpy pandas: ~2s")
print("  â†’ 20x plus rapide! âš¡")
````

````python
from pathlib import Path
import shutil

# 1. CrÃ©er la structure
project_root = Path('data-pipeline')
project_root.mkdir(exist_ok=True)

# Structure des dossiers
folders = [
    project_root / 'src' / 'data_pipeline',
    project_root / 'tests',
    project_root / 'docs',
]

for folder in folders:
    folder.mkdir(parents=True, exist_ok=True)

# 2. pyproject.toml
pyproject_content = '''
[build-system]
requires = ["setuptools>=68.0", "wheel"]
build-backend = "setuptools.build_meta"

[project]
name = "data-pipeline"
version = "0.1.0"
description = "A data pipeline project"
readme = "README.md"
requires-python = ">=3.11"
license = {text = "MIT"}
authors = [
    {name = "Your Name", email = "your@email.com"},
]

dependencies = [
    "pandas>=2.0.0",
    "requests>=2.31.0",
    "python-dotenv>=1.0.0",
]

[project.optional-dependencies]
dev = [
    "pytest>=7.4.0",
    "black>=23.0.0",
    "ruff>=0.1.0",
]

[tool.black]
line-length = 100

[tool.ruff]
line-length = 100
'''

(project_root / 'pyproject.toml').write_text(pyproject_content.strip())

# 3. Module simple
init_content = '''
"""Data Pipeline Package."""
__version__ = "0.1.0"
'''

(project_root / 'src' / 'data_pipeline' / '__init__.py').write_text(init_content.strip())

core_content = '''
"""Core functionality for data pipeline."""
import pandas as pd
import requests
from pathlib import Path

def load_csv(filepath: str | Path) -> pd.DataFrame:
    """Load data from CSV file."""
    return pd.read_csv(filepath)

def fetch_data(url: str) -> dict:
    """Fetch data from API."""
    response = requests.get(url)
    response.raise_for_status()
    return response.json()

def transform_data(df: pd.DataFrame) -> pd.DataFrame:
    """Transform data."""
    # Example transformation
    return df.dropna()
'''

(project_root / 'src' / 'data_pipeline' / 'core.py').write_text(core_content.strip())

# README.md
readme_content = '''
# Data Pipeline

A data processing pipeline project.

## Installation

```bash
# Create virtual environment
python -m venv venv
source venv/bin/activate  # Linux/macOS
# or
venv\\Scripts\\activate  # Windows

# Install in editable mode with dev dependencies
pip install -e ".[dev]"
```

## Usage

```python
from data_pipeline.core import load_csv, transform_data

df = load_csv("data.csv")
df_clean = transform_data(df)
```

## Development

```bash
# Run tests
pytest

# Format code
black src/ tests/

# Lint
ruff check src/ tests/
```
'''

(project_root / 'README.md').write_text(readme_content.strip())

# .gitignore
gitignore_content = '''
# Python
__pycache__/
*.py[cod]
*.so
.Python

# Virtual environments
venv/
.venv/
env/

# Environment variables
.env
.env.local

# Testing
.pytest_cache/
.coverage
htmlcov/

# IDE
.vscode/
.idea/
*.swp

# Build
dist/
build/
*.egg-info/
'''

(project_root / '.gitignore').write_text(gitignore_content.strip())

# Test file
test_content = '''
"""Tests for core module."""
import pandas as pd
from data_pipeline.core import transform_data

def test_transform_data():
    """Test transform_data function."""
    df = pd.DataFrame({
        'a': [1, 2, None, 4],
        'b': [5, None, 7, 8]
    })
    
    result = transform_data(df)
    
    assert len(result) == 2  # Only rows without NaN
    assert result.isna().sum().sum() == 0
'''

(project_root / 'tests' / 'test_core.py').write_text(test_content.strip())
(project_root / 'tests' / '__init__.py').write_text('')

# Afficher la structure
print("âœ“ Projet crÃ©Ã©!")
print("\nStructure:")
print("=" * 60)

def print_tree(directory, prefix="", max_depth=3, current_depth=0):
    if current_depth >= max_depth:
        return
    
    contents = sorted(directory.iterdir(), key=lambda x: (not x.is_dir(), x.name))
    
    for i, path in enumerate(contents):
        is_last = i == len(contents) - 1
        print(f"{prefix}{'â””â”€â”€ ' if is_last else 'â”œâ”€â”€ '}{path.name}")
        
        if path.is_dir() and not path.name.startswith('.'):
            extension = "    " if is_last else "â”‚   "
            print_tree(path, prefix + extension, max_depth, current_depth + 1)

print_tree(project_root)

print("\n4. Pour installer:")
print("  cd data-pipeline")
print("  python -m venv venv")
print("  source venv/bin/activate")
print("  pip install -e '.[dev]'")

# Nettoyer (pour le notebook)
print("\nâœ“ Nettoyage du projet de test...")
shutil.rmtree(project_root)
print("âœ“ TerminÃ©")
````

### Exercice 2: Script d'initialisation de projet

````python
# Ã€ complÃ©ter
# CrÃ©er un script qui:
# 1. Demande le nom du projet
# 2. CrÃ©e la structure complÃ¨te
# 3. GÃ©nÃ¨re pyproject.toml, README, .gitignore
# 4. Initialise git
# 5. CrÃ©e et active un venv
````

### Solution Exercice 2

````python
from pathlib import Path
import subprocess
from datetime import datetime

def create_python_project(project_name: str, author_name: str = "Your Name", author_email: str = "your@email.com"):
    """CrÃ©e un projet Python complet."""
    
    print(f"Creating project: {project_name}")
    print("=" * 60)
    
    # Nom du package (snake_case)
    package_name = project_name.lower().replace("-", "_")
    
    # CrÃ©er la structure
    root = Path(project_name)
    root.mkdir(exist_ok=True)
    
    folders = [
        root / "src" / package_name,
        root / "tests",
        root / "docs",
    ]
    
    for folder in folders:
        folder.mkdir(parents=True, exist_ok=True)
        print(f"âœ“ Created {folder}")
    
    # pyproject.toml
    pyproject = f'''
[build-system]
requires = ["setuptools>=68.0", "wheel"]
build-backend = "setuptools.build_meta"

[project]
name = "{project_name}"
version = "0.1.0"
description = "A Python project"
readme = "README.md"
requires-python = ">=3.11"
license = {{text = "MIT"}}
authors = [
    {{name = "{author_name}", email = "{author_email}"}},
]

dependencies = []

[project.optional-dependencies]
dev = [
    "pytest>=7.4.0",
    "pytest-cov>=4.1.0",
    "black>=23.0.0",
    "ruff>=0.1.0",
    "mypy>=1.7.0",
]

[tool.black]
line-length = 100
target-version = ["py311"]

[tool.ruff]
line-length = 100
target-version = "py311"

[tool.pytest.ini_options]
testpaths = ["tests"]
python_files = ["test_*.py"]
addopts = "-v --cov={package_name}"
'''
    (root / "pyproject.toml").write_text(pyproject.strip())
    print("âœ“ Created pyproject.toml")
    
    # README.md
    readme = f'''
# {project_name}

Created on {datetime.now().strftime("%Y-%m-%d")}

## Installation

```bash
python -m venv venv
source venv/bin/activate
pip install -e ".[dev]"
```

## Usage

```python
import {package_name}
```

## Development

```bash
pytest
black src/ tests/
ruff check src/ tests/
```
'''
    (root / "README.md").write_text(readme.strip())
    print("âœ“ Created README.md")
    
    # .gitignore
    gitignore = '''
# Python
__pycache__/
*.py[cod]
*.so
.Python
venv/
.venv/
env/
.env
.pytest_cache/
.coverage
htmlcov/
dist/
build/
*.egg-info/
.vscode/
.idea/
'''
    (root / ".gitignore").write_text(gitignore.strip())
    print("âœ“ Created .gitignore")
    
    # __init__.py
    init_py = f'''
"""{project_name} package."""
__version__ = "0.1.0"
'''
    (root / "src" / package_name / "__init__.py").write_text(init_py.strip())
    (root / "tests" / "__init__.py").write_text("")
    print("âœ“ Created package files")
    
    # Test file
    test_file = f'''
"""Test {package_name} package."""
import {package_name}

def test_version():
    """Test version is defined."""
    assert {package_name}.__version__ == "0.1.0"
'''
    (root / "tests" / f"test_{package_name}.py").write_text(test_file.strip())
    print("âœ“ Created test file")
    
    # Git init
    try:
        subprocess.run(["git", "init"], cwd=root, check=True, capture_output=True)
        print("âœ“ Initialized git repository")
    except:
        print("âš  Git not available")
    
    print("\n" + "=" * 60)
    print("Project created successfully!")
    print("\nNext steps:")
    print(f"  cd {project_name}")
    print("  python -m venv venv")
    print("  source venv/bin/activate")
    print("  pip install -e '.[dev]'")
    print("  pytest")
    
    return root

# Utilisation
project = create_python_project(
    "my-awesome-project",
    author_name="John Doe",
    author_email="john@example.com"
)

# Nettoyer
import shutil
shutil.rmtree(project)
print("\nâœ“ Test project cleaned up")
````
