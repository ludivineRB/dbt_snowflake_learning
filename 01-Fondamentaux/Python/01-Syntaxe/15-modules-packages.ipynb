## 1. Qu'est-ce qu'un module?

Un **module** est simplement un fichier Python (`.py`) contenant du code (fonctions, classes, variables).

### Avantages des modules:

1. **Reutilisabilite**: partager du code entre plusieurs programmes
2. **Organisation**: separer le code en fichiers thematiques
3. **Namespace**: eviter les conflits de noms
4. **Maintenance**: code plus facile a maintenir

````python
# Exemple: utiliser un module de la bibliotheque standard
import math

print(f"Pi: {math.pi}")
print(f"Racine carree de 16: {math.sqrt(16)}")
print(f"Cosinus de 0: {math.cos(0)}")

# Que contient le module math?
print(f"\nNombre d'attributs dans math: {len(dir(math))}")
print(f"\nQuelques fonctions: {[x for x in dir(math) if not x.startswith('_')][:10]}")
````

## 3. __name__ et __main__

La variable speciale `__name__` indique:
- `"__main__"` si le fichier est execute directement
- Le nom du module si le fichier est importe

````python
# Dans un notebook, __name__ est toujours '__main__'
print(f"__name__ = {__name__}")

# Pattern classique dans les fichiers .py
def ma_fonction():
    print("Fonction appelee")

if __name__ == "__main__":
    # Ce code ne s'execute que si le fichier est lance directement
    # Pas si le fichier est importe comme module
    print("Fichier execute directement")
    ma_fonction()
````

### Exemple pratique

Voici comment structurer un fichier Python reutilisable:

````python
# Contenu d'un fichier 'calculatrice.py' (exemple)
exemple_module = '''
"""Module calculatrice - operations mathematiques simples"""

def addition(a, b):
    """Additionne deux nombres"""
    return a + b

def multiplication(a, b):
    """Multiplie deux nombres"""
    return a * b

def division(a, b):
    """Divise deux nombres"""
    if b == 0:
        raise ValueError("Division par zero")
    return a / b

# Code de test / demonstration
if __name__ == "__main__":
    # Ce code ne s'execute que si on lance directement calculatrice.py
    print("Test du module calculatrice")
    print(f"5 + 3 = {addition(5, 3)}")
    print(f"5 * 3 = {multiplication(5, 3)}")
    print(f"10 / 2 = {division(10, 2)}")
'''

print("Exemple de structure de module:")
print(exemple_module)
````

## 4. Creer son propre module

Creons un module simple dans un fichier temporaire.

````python
import os
import sys
import tempfile

# Creer un repertoire temporaire pour nos modules
temp_dir = tempfile.mkdtemp()
print(f"Repertoire temporaire: {temp_dir}")

# Ajouter le repertoire au path Python
sys.path.insert(0, temp_dir)

# Creer un module 'utilitaires.py'
module_path = os.path.join(temp_dir, 'utilitaires.py')
with open(module_path, 'w') as f:
    f.write('''
"""Module utilitaires - fonctions diverses"""

def saluer(nom):
    """Retourne un message de salutation"""
    return f"Bonjour {nom}!"

def carre(x):
    """Calcule le carre d'un nombre"""
    return x ** 2

def est_pair(n):
    """Verifie si un nombre est pair"""
    return n % 2 == 0

# Variable du module
VERSION = "1.0.0"

if __name__ == "__main__":
    print("Test du module utilitaires")
    print(saluer("Alice"))
    print(f"Carre de 5: {carre(5)}")
    print(f"8 est pair: {est_pair(8)}")
''')

print(f"Module cree: {module_path}")
````

````python
# Importer notre module personnalise
import utilitaires

print(f"Version du module: {utilitaires.VERSION}")
print(f"\n{utilitaires.saluer('Bob')}")
print(f"Carre de 7: {utilitaires.carre(7)}")
print(f"15 est pair: {utilitaires.est_pair(15)}")

# Documentation du module
print(f"\nDocumentation: {utilitaires.__doc__}")
print(f"\nFonctions disponibles: {[x for x in dir(utilitaires) if not x.startswith('_')]}")
````

````python
# Import selectif
from utilitaires import saluer, carre

print(saluer('Charlie'))
print(f"Carre de 9: {carre(9)}")
````

## 5. Structure d'un package

Un **package** est un repertoire contenant:
1. Un fichier `__init__.py` (peut etre vide)
2. Un ou plusieurs modules Python

```
mon_package/
    __init__.py
    module1.py
    module2.py
    sous_package/
        __init__.py
        module3.py
```

````python
# Creer une structure de package
package_dir = os.path.join(temp_dir, 'mon_package')
os.makedirs(package_dir, exist_ok=True)

# __init__.py
with open(os.path.join(package_dir, '__init__.py'), 'w') as f:
    f.write('''
"""Mon package - ensemble de modules utiles"""

__version__ = "1.0.0"
__author__ = "Formation Data Engineer"

# Importer des fonctions pour les rendre accessibles au niveau du package
from .mathematiques import addition, multiplication
from .texte import majuscules, compter_mots

__all__ = ['addition', 'multiplication', 'majuscules', 'compter_mots']
''')

# Module mathematiques.py
with open(os.path.join(package_dir, 'mathematiques.py'), 'w') as f:
    f.write('''
"""Module mathematiques - operations mathematiques"""

def addition(a, b):
    """Additionne deux nombres"""
    return a + b

def soustraction(a, b):
    """Soustrait deux nombres"""
    return a - b

def multiplication(a, b):
    """Multiplie deux nombres"""
    return a * b

def division(a, b):
    """Divise deux nombres"""
    if b == 0:
        raise ValueError("Division par zero")
    return a / b
''')

# Module texte.py
with open(os.path.join(package_dir, 'texte.py'), 'w') as f:
    f.write('''
"""Module texte - operations sur les chaines"""

def majuscules(texte):
    """Convertit en majuscules"""
    return texte.upper()

def minuscules(texte):
    """Convertit en minuscules"""
    return texte.lower()

def compter_mots(texte):
    """Compte le nombre de mots"""
    return len(texte.split())

def inverser(texte):
    """Inverse une chaine"""
    return texte[::-1]
''')

print(f"Package cree dans: {package_dir}")
print(f"\nStructure:")
print(f"  mon_package/")
print(f"    __init__.py")
print(f"    mathematiques.py")
print(f"    texte.py")
````

````python
# Importer le package
import mon_package

print(f"Version: {mon_package.__version__}")
print(f"Auteur: {mon_package.__author__}")

# Utiliser les fonctions importees dans __init__.py
print(f"\n5 + 3 = {mon_package.addition(5, 3)}")
print(f"5 * 3 = {mon_package.multiplication(5, 3)}")
print(f"'python' en majuscules: {mon_package.majuscules('python')}")
print(f"Nombre de mots dans 'Hello World': {mon_package.compter_mots('Hello World')}")
````

````python
# Importer des modules specifiques
from mon_package import mathematiques, texte

print(f"Division: {mathematiques.division(10, 2)}")
print(f"Soustraction: {mathematiques.soustraction(10, 3)}")
print(f"Minuscules: {texte.minuscules('PYTHON')}")
print(f"Inverse: {texte.inverser('Python')}")
````

````python
# Import selectif depuis un sous-module
from mon_package.mathematiques import addition, division
from mon_package.texte import inverser

print(f"7 + 8 = {addition(7, 8)}")
print(f"20 / 4 = {division(20, 4)}")
print(f"'Bonjour' inverse: {inverser('Bonjour')}")
````

## 6. Imports relatifs vs absolus

### Imports absolus

Utilisent le chemin complet depuis la racine du package.

```python
from mon_package.mathematiques import addition
```

### Imports relatifs

Utilisent `.` pour le repertoire courant et `..` pour le parent.

```python
from .mathematiques import addition  # Meme niveau
from ..autre_package import fonction  # Niveau parent
```

````python
# Exemple d'imports relatifs dans un package
exemple_relatif = '''
# Dans mon_package/texte.py
from .mathematiques import addition  # Import relatif

def compter_et_ajouter(texte, n):
    nb_mots = len(texte.split())
    return addition(nb_mots, n)  # Utilise la fonction du module voisin
'''

print("Exemple d'import relatif:")
print(exemple_relatif)

print("\nRecommandation:")
print("- Utilisez les imports ABSOLUS pour plus de clarte")
print("- Les imports relatifs sont utiles dans les gros packages")
````

## 8. pip: installer des packages

**pip** est l'outil standard pour installer des packages Python depuis PyPI (Python Package Index).

### Commandes principales

```bash

pip list

pip show requests
```

````python
# Lister quelques packages installes
import subprocess

try:
    result = subprocess.run(['pip', 'list'], capture_output=True, text=True, timeout=5)
    lignes = result.stdout.split('\n')[:10]
    print("Quelques packages installes:")
    for ligne in lignes:
        print(ligne)
except Exception as e:
    print(f"Impossible de lister les packages: {e}")
````

## Pieges courants

### 1. Imports circulaires

````python
# PIEGE: imports circulaires
exemple_circulaire = '''
# fichier a.py
from b import fonction_b

def fonction_a():
    return fonction_b()

# fichier b.py
from a import fonction_a  # ERREUR: import circulaire!

def fonction_b():
    return fonction_a()
'''

print("Exemple d'import circulaire (A EVITER):")
print(exemple_circulaire)

# SOLUTION 1: Reorganiser le code
solution1 = '''
# Deplacer les fonctions dans un troisieme module
# fichier utils.py
def fonction_commune():
    pass

# fichier a.py
from utils import fonction_commune

# fichier b.py
from utils import fonction_commune
'''

# SOLUTION 2: Import local (dans la fonction)
solution2 = '''
# fichier a.py
def fonction_a():
    from b import fonction_b  # Import local
    return fonction_b()
'''

print("\nSolution 1 (reorganisation):")
print(solution1)
print("\nSolution 2 (import local):")
print(solution2)
````

### 2. Shadowing (masquage de modules)

````python
# PIEGE: nommer un fichier comme un module standard
exemple_shadowing = '''
# NE JAMAIS FAIRE CA!
# Si vous creez un fichier nomme 'random.py' dans votre projet,
# il masquera le module random de la stdlib!

# Votre fichier random.py
def ma_fonction():
    pass

# Dans un autre fichier
import random  # ERREUR: importe VOTRE random.py, pas le module stdlib!
print(random.randint(1, 10))  # AttributeError: pas de randint
'''

print("ATTENTION: Shadowing (masquage):")
print(exemple_shadowing)

print("\nModules stdlib a NE PAS utiliser comme noms:")
print("- random, math, os, sys, json, datetime, etc.")
print("- Verifiez toujours: https://docs.python.org/3/py-modindex.html")
````

### 3. Import wildcard (import *)

````python
# PIEGE: import * peut causer des conflits
print("PROBLEME avec 'from module import *':")
print("")
print("1. On ne sait pas ce qui est importe")
print("2. Risque de conflits de noms")
print("3. Code difficile a lire et maintenir")
print("4. Ralentit certains outils (linters, IDE)")

exemple_conflit = '''
from math import *  # Importe tout
from numpy import *  # Importe tout aussi

# Laquelle de ces fonctions est utilisee?
result = sqrt(2)  # math.sqrt ou numpy.sqrt?
'''

print("\nExemple de conflit:")
print(exemple_conflit)

print("\nSOLUTION: Import explicite")
print("import math")
print("import numpy as np")
print("result = math.sqrt(2)  # Clair!")
````

## Solutions

### Solution Exercice 1

````python
# Creer le module strings_utils.py
strings_utils_path = os.path.join(temp_dir, 'strings_utils.py')
with open(strings_utils_path, 'w') as f:
    f.write('''
"""Module strings_utils - utilitaires pour les chaines"""

def compter_voyelles(texte):
    """Compte le nombre de voyelles dans un texte"""
    voyelles = 'aeiouyAEIOUY'
    return sum(1 for char in texte if char in voyelles)

def est_palindrome(texte):
    """Verifie si un texte est un palindrome"""
    # Nettoyer: minuscules et sans espaces
    texte_propre = ''.join(texte.lower().split())
    return texte_propre == texte_propre[::-1]

def premier_mot(texte):
    """Retourne le premier mot d'un texte"""
    mots = texte.split()
    return mots[0] if mots else ''

if __name__ == "__main__":
    # Tests
    print("Tests du module strings_utils")
    print(f"Voyelles dans 'bonjour': {compter_voyelles('bonjour')}")
    print(f"'radar' est palindrome: {est_palindrome('radar')}")
    print(f"'python' est palindrome: {est_palindrome('python')}")
    print(f"Premier mot de 'Hello World': {premier_mot('Hello World')}")
''')

# Importer et tester
import strings_utils

print("Tests:")
print(f"Voyelles dans 'Python': {strings_utils.compter_voyelles('Python')}")
print(f"'Kayak' est palindrome: {strings_utils.est_palindrome('Kayak')}")
print(f"'Elu par cette crapule' est palindrome: {strings_utils.est_palindrome('Elu par cette crapule')}")
print(f"Premier mot: {strings_utils.premier_mot('Formation Data Engineer')}")
````
