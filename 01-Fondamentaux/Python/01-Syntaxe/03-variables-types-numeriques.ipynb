## 1. Variables et Affectation

### Qu'est-ce qu'une variable ?

Une **variable** est un nom qui fait r√©f√©rence √† une valeur stock√©e en m√©moire.

En Python :
- Pas de d√©claration de type n√©cessaire
- L'op√©rateur `=` effectue l'affectation
- La variable est cr√©√©e lors de la premi√®re affectation

````python
# Affectation simple
age = 25
prenom = "Alice"
taille = 1.75
est_etudiant = True

print(f"Pr√©nom : {prenom}")
print(f"√Çge : {age}")
print(f"Taille : {taille} m")
print(f"√âtudiant : {est_etudiant}")
````

### Affectation multiple

````python
# Affecter plusieurs variables en une ligne
x, y, z = 10, 20, 30
print(f"x={x}, y={y}, z={z}")

# Affecter la m√™me valeur √† plusieurs variables
a = b = c = 100
print(f"a={a}, b={b}, c={c}")

# √âchanger deux variables
x, y = 5, 10
print(f"Avant : x={x}, y={y}")
x, y = y, x  # Swap √©l√©gant en Python
print(f"Apr√®s : x={x}, y={y}")
````

### Affectation augment√©e

````python
# Op√©rateurs d'affectation augment√©e
compteur = 10

compteur += 5   # √âquivalent √† : compteur = compteur + 5
print(f"Apr√®s +=5 : {compteur}")

compteur -= 3   # √âquivalent √† : compteur = compteur - 3
print(f"Apr√®s -=3 : {compteur}")

compteur *= 2   # √âquivalent √† : compteur = compteur * 2
print(f"Apr√®s *=2 : {compteur}")

compteur /= 4   # √âquivalent √† : compteur = compteur / 4
print(f"Apr√®s /=4 : {compteur}")

# Autres op√©rateurs augment√©s
n = 17
n //= 5  # Division enti√®re
print(f"17 //= 5 : {n}")

n = 17
n %= 5   # Modulo
print(f"17 %= 5 : {n}")

n = 2
n **= 10  # Puissance
print(f"2 **= 10 : {n}")
````

## 2. Typage Dynamique

Python est un langage √† **typage dynamique** :
- Le type est d√©termin√© √† l'ex√©cution
- Une variable peut changer de type au cours du programme
- Pas besoin de d√©clarer le type explicitement

````python
# La variable change de type dynamiquement
variable = 42
print(f"Type : {type(variable)}, Valeur : {variable}")

variable = "Maintenant c'est une cha√Æne"
print(f"Type : {type(variable)}, Valeur : {variable}")

variable = 3.14
print(f"Type : {type(variable)}, Valeur : {variable}")

variable = True
print(f"Type : {type(variable)}, Valeur : {variable}")
````

### Typage dynamique vs typage statique

**Python (typage dynamique) :**
```python
x = 10        # Python d√©duit que c'est un int
x = "texte"   # OK : x change de type
```

**Java (typage statique) :**
```java
int x = 10;
x = "texte";  // ERREUR : x est d√©clar√© comme int
```

## 3. Types Num√©riques

### 3.1. Entiers (int)

En Python 3, les entiers ont une **pr√©cision illimit√©e** (limit√©s uniquement par la m√©moire).

````python
# Entiers standards
age = 25
population = 67000000
negatif = -42

print(f"Type de age : {type(age)}")
print(f"Valeur : {age}")

# Entiers TR√àS grands (pr√©cision illimit√©e)
grand_nombre = 123456789012345678901234567890
print(f"\nGrand nombre : {grand_nombre}")
print(f"Type : {type(grand_nombre)}")

# Calcul avec de tr√®s grands nombres
resultat = grand_nombre ** 2
print(f"\nCarr√© : {resultat}")
````

````python
# Diff√©rentes bases num√©riques
decimal = 42        # Base 10 (d√©cimal)
binaire = 0b101010  # Base 2 (binaire)
octal = 0o52        # Base 8 (octal)
hexa = 0x2A         # Base 16 (hexad√©cimal)

print(f"D√©cimal : {decimal}")
print(f"Binaire 0b101010 : {binaire}")
print(f"Octal 0o52 : {octal}")
print(f"Hexad√©cimal 0x2A : {hexa}")
print(f"\nTous repr√©sentent : {decimal}")

# Conversion vers d'autres bases
n = 42
print(f"\n{n} en binaire : {bin(n)}")
print(f"{n} en octal : {oct(n)}")
print(f"{n} en hexad√©cimal : {hex(n)}")
````

````python
# S√©parateurs de milliers (Python 3.6+)
population = 67_000_000  # Plus lisible
prix = 1_234_567.89

print(f"Population : {population:,}")
print(f"Prix : {prix:,.2f}")
````

### 3.2. Flottants (float)

Les **flottants** repr√©sentent les nombres d√©cimaux (virgule flottante IEEE 754).

````python
# Nombres √† virgule flottante
pi = 3.14159
temperature = -15.5
scientifique = 1.5e-10  # Notation scientifique : 1.5 √ó 10^-10

print(f"Pi : {pi}")
print(f"Temp√©rature : {temperature}¬∞C")
print(f"Notation scientifique : {scientifique}")
print(f"Type : {type(pi)}")
````

````python
# Notation scientifique
vitesse_lumiere = 3e8   # 3 √ó 10^8 m/s
masse_electron = 9.1e-31  # 9.1 √ó 10^-31 kg

print(f"Vitesse de la lumi√®re : {vitesse_lumiere} m/s")
print(f"Masse de l'√©lectron : {masse_electron} kg")
````

````python
# Valeurs sp√©ciales
infini = float('inf')
moins_infini = float('-inf')
pas_un_nombre = float('nan')  # Not a Number

print(f"Infini : {infini}")
print(f"Moins infini : {moins_infini}")
print(f"NaN : {pas_un_nombre}")

# Test
import math
print(f"\nEst infini ? {math.isinf(infini)}")
print(f"Est NaN ? {math.isnan(pas_un_nombre)}")
````

### 3.3. Nombres Complexes (complex)

Python supporte nativement les **nombres complexes** (partie r√©elle + partie imaginaire).

````python
# Nombres complexes (a + bj)
z1 = 3 + 4j
z2 = complex(2, -5)  # 2 - 5j

print(f"z1 = {z1}")
print(f"z2 = {z2}")
print(f"Type : {type(z1)}")

# Acc√®s aux parties r√©elle et imaginaire
print(f"\nPartie r√©elle de z1 : {z1.real}")
print(f"Partie imaginaire de z1 : {z1.imag}")

# Conjugu√©
print(f"\nConjugu√© de z1 : {z1.conjugate()}")

# Op√©rations
print(f"\nz1 + z2 = {z1 + z2}")
print(f"z1 * z2 = {z1 * z2}")

# Module (valeur absolue)
print(f"\n|z1| = {abs(z1)}")
````

## 4. Op√©rations Arithm√©tiques

````python
# Op√©rations de base
a = 10
b = 3

print(f"a = {a}, b = {b}\n")
print(f"Addition : a + b = {a + b}")
print(f"Soustraction : a - b = {a - b}")
print(f"Multiplication : a * b = {a * b}")
print(f"Division : a / b = {a / b}")
print(f"Division enti√®re : a // b = {a // b}")
print(f"Modulo : a % b = {a % b}")
print(f"Puissance : a ** b = {a ** b}")
````

````python
# Division : TOUJOURS un float en Python 3
print(f"10 / 3 = {10 / 3}")    # 3.3333... (float)
print(f"10 / 2 = {10 / 2}")    # 5.0 (float, m√™me si divisible)

# Division enti√®re : r√©sultat tronqu√©
print(f"\n10 // 3 = {10 // 3}")  # 3 (int)
print(f"10 // 2 = {10 // 2}")    # 5 (int)
print(f"-10 // 3 = {-10 // 3}")  # -4 (arrondi vers -‚àû, pas vers 0)

# Modulo : reste de la division
print(f"\n10 % 3 = {10 % 3}")   # 1
print(f"10 % 2 = {10 % 2}")     # 0 (divisible)
````

````python
# Priorit√© des op√©rations (comme en math√©matiques)
resultat = 2 + 3 * 4
print(f"2 + 3 * 4 = {resultat}")  # 14 (pas 20)

resultat = (2 + 3) * 4
print(f"(2 + 3) * 4 = {resultat}")  # 20

# Ordre : ** > */ > +-
resultat = 2 + 3 * 4 ** 2
print(f"2 + 3 * 4 ** 2 = {resultat}")  # 2 + 3*16 = 50
````

## 5. Bool√©ens (bool)

Le type **bool** ne peut prendre que deux valeurs : `True` ou `False`.

````python
# Valeurs bool√©ennes
vrai = True
faux = False

print(f"Type : {type(vrai)}")
print(f"True : {vrai}")
print(f"False : {faux}")

# bool est une sous-classe de int
print(f"\nTrue == 1 : {True == 1}")
print(f"False == 0 : {False == 0}")
print(f"True + True : {True + True}")  # 2
print(f"True * 5 : {True * 5}")        # 5
````

````python
# Op√©rateurs de comparaison retournent des bool
age = 18

print(f"age == 18 : {age == 18}")
print(f"age != 18 : {age != 18}")
print(f"age > 16 : {age > 16}")
print(f"age >= 18 : {age >= 18}")
print(f"age < 21 : {age < 21}")
print(f"age <= 18 : {age <= 18}")
````

### Valeurs Truthy et Falsy

En Python, toute valeur peut √™tre √©valu√©e comme bool√©en dans un contexte bool√©en.

**Valeurs Falsy (√©valu√©es √† False) :**
- `False`
- `None`
- `0`, `0.0`, `0j`
- S√©quences vides : `''`, `[]`, `()`, `{}`

**Tout le reste est Truthy (√©valu√© √† True)**

````python
# Conversion en bool√©en avec bool()
print("Valeurs FALSY :")
print(f"bool(False) = {bool(False)}")
print(f"bool(None) = {bool(None)}")
print(f"bool(0) = {bool(0)}")
print(f"bool(0.0) = {bool(0.0)}")
print(f"bool('') = {bool('')}")  # Cha√Æne vide
print(f"bool([]) = {bool([])}")

print("\nValeurs TRUTHY :")
print(f"bool(True) = {bool(True)}")
print(f"bool(1) = {bool(1)}")
print(f"bool(-1) = {bool(-1)}")
print(f"bool('texte') = {bool('texte')}")
print(f"bool([1, 2]) = {bool([1, 2])}")
````

````python
# Op√©rateurs logiques
a = True
b = False

print(f"a and b : {a and b}")
print(f"a or b : {a or b}")
print(f"not a : {not a}")
print(f"not b : {not b}")

# Court-circuit
print(f"\nTrue or ERREUR : {True or 1/0}")  # N'√©value pas 1/0
# print(f"False and ERREUR : {False and 1/0}")  # N'√©value pas 1/0
````

## 6. None : L'Absence de Valeur

`None` est un objet sp√©cial qui repr√©sente l'absence de valeur.

````python
# None : singleton (une seule instance)
rien = None

print(f"Type : {type(rien)}")
print(f"Valeur : {rien}")

# Test avec 'is' (identit√©)
autre_rien = None
print(f"\nrien is None : {rien is None}")
print(f"rien is autre_rien : {rien is autre_rien}")

# None est Falsy
print(f"\nbool(None) : {bool(None)}")

if not rien:
    print("None est √©valu√© √† False")
````

````python
# Utilisation typique : valeur par d√©faut
def obtenir_nom(utilisateur=None):
    """Retourne le nom de l'utilisateur ou 'Anonyme'."""
    if utilisateur is None:
        return "Anonyme"
    return utilisateur

print(obtenir_nom("Alice"))
print(obtenir_nom())
````

## 7. V√©rification et Conversion de Types

### type() et isinstance()

````python
# type() : retourne le type exact
x = 42
y = 3.14
z = "texte"

print(f"type(42) : {type(x)}")
print(f"type(3.14) : {type(y)}")
print(f"type('texte') : {type(z)}")

# Comparaison de types
print(f"\ntype(x) == int : {type(x) == int}")
print(f"type(y) == float : {type(y) == float}")
````

````python
# isinstance() : v√©rifie si un objet est d'un type donn√©
# (pr√©f√©r√© √† type() car g√®re l'h√©ritage)

x = 42

print(f"isinstance(x, int) : {isinstance(x, int)}")
print(f"isinstance(x, float) : {isinstance(x, float)}")
print(f"isinstance(x, (int, float)) : {isinstance(x, (int, float))}")  # Tuple de types

# bool est une sous-classe de int
b = True
print(f"\nisinstance(True, bool) : {isinstance(b, bool)}")
print(f"isinstance(True, int) : {isinstance(b, int)}")  # True aussi !
````

### Conversions de types (casting)

````python
# int() : conversion en entier
print("Conversion en int :")
print(f"int(3.14) : {int(3.14)}")        # Tronque (ne arrondit pas)
print(f"int('42') : {int('42')}")        # Depuis une cha√Æne
print(f"int('1010', 2) : {int('1010', 2)}")  # Depuis binaire
print(f"int(True) : {int(True)}")        # 1
print(f"int(False) : {int(False)}")      # 0
````

````python
# float() : conversion en flottant
print("Conversion en float :")
print(f"float(42) : {float(42)}")
print(f"float('3.14') : {float('3.14')}")
print(f"float('1e-3') : {float('1e-3')}")
print(f"float(True) : {float(True)}")
````

````python
# str() : conversion en cha√Æne
print("Conversion en str :")
print(f"str(42) : {str(42)}")
print(f"str(3.14) : {str(3.14)}")
print(f"str(True) : {str(True)}")
print(f"str(None) : {str(None)}")
````

````python
# bool() : conversion en bool√©en
print("Conversion en bool :")
print(f"bool(1) : {bool(1)}")
print(f"bool(0) : {bool(0)}")
print(f"bool('texte') : {bool('texte')}")
print(f"bool('') : {bool('')}")
print(f"bool(None) : {bool(None)}")
````

````python
# Conversions qui √©chouent
try:
    int("texte")  # ValueError
except ValueError as e:
    print(f"Erreur : {e}")

try:
    float("3.14.15")  # ValueError
except ValueError as e:
    print(f"Erreur : {e}")
````

## 8. Port√©e des Variables (Scope)

### Port√©e locale vs globale

````python
# Variable globale
x = 100

def ma_fonction():
    # Variable locale (masque la globale)
    x = 10
    print(f"Dans la fonction, x = {x}")

ma_fonction()
print(f"En dehors, x = {x}")  # La globale n'a pas chang√©
````

````python
# Utiliser une variable globale (lecture)
y = 50

def lire_globale():
    print(f"Je peux lire y : {y}")

lire_globale()
````

````python
# Modifier une variable globale (mot-cl√© global)
compteur = 0

def incrementer():
    global compteur  # D√©clare qu'on utilise la globale
    compteur += 1
    print(f"Compteur dans fonction : {compteur}")

print(f"Avant : {compteur}")
incrementer()
incrementer()
print(f"Apr√®s : {compteur}")
````

## 9. Pi√®ges Courants

### Pi√®ge 1 : Division enti√®re vs division flottante

````python
# ‚ö†Ô∏è En Python 3, / retourne TOUJOURS un float
print(f"5 / 2 = {5 / 2}")   # 2.5 (float)
print(f"5 // 2 = {5 // 2}") # 2 (int)

# En Python 2, 5 / 2 donnait 2 (division enti√®re par d√©faut)
# Utilisez // si vous voulez explicitement la division enti√®re
````

### Pi√®ge 2 : Pr√©cision des flottants (0.1 + 0.2 ‚â† 0.3)

````python
# ‚ö†Ô∏è Probl√®me de pr√©cision flottante IEEE 754
resultat = 0.1 + 0.2
print(f"0.1 + 0.2 = {resultat}")
print(f"0.1 + 0.2 == 0.3 : {resultat == 0.3}")  # False !

# Repr√©sentation binaire approximative
print(f"\nRepr√©sentation pr√©cise : {resultat:.20f}")

# Solution 1 : Comparaison avec tol√©rance
epsilon = 1e-10
print(f"\nAvec tol√©rance : {abs(resultat - 0.3) < epsilon}")

# Solution 2 : Module math.isclose()
import math
print(f"math.isclose() : {math.isclose(resultat, 0.3)}")

# Solution 3 : Module decimal pour pr√©cision exacte
from decimal import Decimal
d1 = Decimal('0.1')
d2 = Decimal('0.2')
print(f"\nAvec Decimal : {d1 + d2}")
print(f"√âgalit√© : {d1 + d2 == Decimal('0.3')}")
````

### Pi√®ge 3 : Comparaison bool avec == au lieu de is

````python
# ‚úÖ BON : Utiliser 'is' pour None, True, False
valeur = None
if valeur is None:
    print("Utilise 'is' pour None")

# ‚ùå Acceptable mais moins idiomatique
if valeur == None:
    print("Fonctionne mais pas recommand√©")

# M√™me chose pour bool
flag = True
if flag is True:
    print("Utilise 'is' pour True/False")
````

### Pi√®ge 4 : Conversion int() tronque, ne arrondit pas

````python
# ‚ö†Ô∏è int() TRONQUE vers z√©ro, ne arrondit PAS
print(f"int(3.9) = {int(3.9)}")   # 3 (pas 4 !)
print(f"int(-3.9) = {int(-3.9)}") # -3 (pas -4 !)

# Pour arrondir, utilisez round()
print(f"\nround(3.9) = {round(3.9)}")   # 4
print(f"round(3.5) = {round(3.5)}")     # 4 (arrondi au pair)
print(f"round(4.5) = {round(4.5)}")     # 4 (arrondi au pair)

# Ou math.floor() / math.ceil()
import math
print(f"\nmath.floor(3.9) = {math.floor(3.9)}")  # 3
print(f"math.ceil(3.1) = {math.ceil(3.1)}")      # 4
````
