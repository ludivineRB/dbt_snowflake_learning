## 1. Cr√©ation de listes

Les listes sont des collections **ordonn√©es** et **mutables** d'√©l√©ments. Elles peuvent contenir des √©l√©ments de types diff√©rents.

````python
# Liste vide
liste_vide1 = []
liste_vide2 = list()

print(f"Liste vide 1: {liste_vide1}")
print(f"Liste vide 2: {liste_vide2}")

# Liste avec des √©l√©ments
nombres = [1, 2, 3, 4, 5]
print(f"\nNombres: {nombres}")

# Liste avec types mixtes
mixte = [1, "Python", 3.14, True, None]
print(f"Mixte: {mixte}")

# Liste √† partir d'une cha√Æne
lettres = list("Python")
print(f"\nLettres: {lettres}")

# Liste √† partir d'un range
nombres_range = list(range(1, 11))
print(f"Range 1-10: {nombres_range}")

nombres_range2 = list(range(0, 20, 2))  # 0 √† 20, pas de 2
print(f"Range 0-20 (pas 2): {nombres_range2}")
````

## 2. Indexation et slicing

Comme les strings, les listes supportent l'indexation et le slicing.

````python
langages = ["Python", "Java", "JavaScript", "C++", "Ruby"]
#            0         1        2            3      4
#           -5        -4       -3           -2     -1

# Indexation
print(f"Premier √©l√©ment: {langages[0]}")
print(f"Dernier √©l√©ment: {langages[-1]}")
print(f"Avant-dernier: {langages[-2]}")

# Modification par index (les listes sont mutables!)
print(f"\nAvant: {langages}")
langages[1] = "Go"
print(f"Apr√®s modification: {langages}")

# Slicing
langages = ["Python", "Java", "JavaScript", "C++", "Ruby"]
print(f"\nlangages[1:4]: {langages[1:4]}")
print(f"langages[:3]: {langages[:3]}")
print(f"langages[2:]: {langages[2:]}")
print(f"langages[:]: {langages[:]}")

# Slicing avec step
print(f"\nlangages[::2]: {langages[::2]}")
print(f"langages[1::2]: {langages[1::2]}")
print(f"langages[::-1]: {langages[::-1]}")  # Inversion

# Modification par slicing
nombres = [1, 2, 3, 4, 5]
print(f"\nAvant: {nombres}")
nombres[1:4] = [20, 30, 40]
print(f"Apr√®s: {nombres}")
````

## 3. M√©thodes d'ajout et de suppression

### Ajout d'√©l√©ments

````python
# append() - ajoute un √©l√©ment √† la fin
fruits = ["pomme", "banane"]
print(f"Initial: {fruits}")

fruits.append("orange")
print(f"Apr√®s append('orange'): {fruits}")

# Attention: append() ajoute l'√©l√©ment tel quel
fruits.append(["kiwi", "mangue"])
print(f"Apr√®s append(['kiwi', 'mangue']): {fruits}")

# extend() - ajoute tous les √©l√©ments d'un it√©rable
fruits = ["pomme", "banane"]
fruits.extend(["kiwi", "mangue"])
print(f"\nApr√®s extend(['kiwi', 'mangue']): {fruits}")

# extend() avec une cha√Æne
lettres = ['a', 'b']
lettres.extend('cd')
print(f"\nLettres apr√®s extend('cd'): {lettres}")

# insert() - ins√®re un √©l√©ment √† une position
nombres = [1, 2, 4, 5]
nombres.insert(2, 3)  # Ins√®re 3 √† l'index 2
print(f"\nApr√®s insert(2, 3): {nombres}")

nombres.insert(0, 0)  # Ins√®re au d√©but
print(f"Apr√®s insert(0, 0): {nombres}")

nombres.insert(100, 6)  # Index trop grand = ajout √† la fin
print(f"Apr√®s insert(100, 6): {nombres}")
````

### Suppression d'√©l√©ments

````python
# remove() - supprime la premi√®re occurrence d'une valeur
langages = ["Python", "Java", "Python", "C++"]
print(f"Initial: {langages}")

langages.remove("Python")
print(f"Apr√®s remove('Python'): {langages}")

# Erreur si l'√©l√©ment n'existe pas
try:
    langages.remove("Ruby")
except ValueError as e:
    print(f"\nErreur: {e}")

# pop() - supprime et retourne l'√©l√©ment √† un index
nombres = [1, 2, 3, 4, 5]
print(f"\nInitial: {nombres}")

dernier = nombres.pop()  # Sans argument = dernier √©l√©ment
print(f"√âl√©ment supprim√©: {dernier}")
print(f"Apr√®s pop(): {nombres}")

deuxieme = nombres.pop(1)  # Supprime l'√©l√©ment √† l'index 1
print(f"√âl√©ment supprim√©: {deuxieme}")
print(f"Apr√®s pop(1): {nombres}")

# del - supprime par index ou slice
nombres = [1, 2, 3, 4, 5]
del nombres[0]  # Supprime le premier √©l√©ment
print(f"\nApr√®s del nombres[0]: {nombres}")

del nombres[1:3]  # Supprime un slice
print(f"Apr√®s del nombres[1:3]: {nombres}")

# clear() - vide la liste
nombres = [1, 2, 3, 4, 5]
nombres.clear()
print(f"\nApr√®s clear(): {nombres}")
````

## 4. M√©thodes de tri et d'organisation

### sort() vs sorted()

````python
# sort() - trie la liste en place (modifie la liste)
nombres = [5, 2, 8, 1, 9]
print(f"Avant sort(): {nombres}")

nombres.sort()
print(f"Apr√®s sort(): {nombres}")

# Tri d√©croissant
nombres.sort(reverse=True)
print(f"Apr√®s sort(reverse=True): {nombres}")

# sorted() - retourne une nouvelle liste tri√©e (ne modifie pas l'original)
nombres_originaux = [5, 2, 8, 1, 9]
nombres_tries = sorted(nombres_originaux)
print(f"\nOriginaux: {nombres_originaux}")
print(f"Tri√©s: {nombres_tries}")

# Tri de cha√Ænes
mots = ["python", "java", "C++", "Ruby"]
mots.sort()
print(f"\nMots tri√©s: {mots}")

# Tri avec cl√© personnalis√©e
mots = ["python", "java", "C++", "Ruby"]
mots.sort(key=len)  # Trie par longueur
print(f"Tri√©s par longueur: {mots}")

mots.sort(key=str.lower)  # Trie sans tenir compte de la casse
print(f"Tri√©s (case-insensitive): {mots}")

# Tri de listes de tuples
etudiants = [("Alice", 18), ("Bob", 20), ("Charlie", 19)]
etudiants.sort(key=lambda x: x[1])  # Trie par √¢ge
print(f"\n√âtudiants tri√©s par √¢ge: {etudiants}")
````

### reverse() et autres m√©thodes

````python
# reverse() - inverse la liste en place
nombres = [1, 2, 3, 4, 5]
print(f"Avant reverse(): {nombres}")

nombres.reverse()
print(f"Apr√®s reverse(): {nombres}")

# Ou avec slicing (cr√©e une nouvelle liste)
nombres = [1, 2, 3, 4, 5]
nombres_inverses = nombres[::-1]
print(f"\nOriginaux: {nombres}")
print(f"Invers√©s (slicing): {nombres_inverses}")

# index() - trouve l'index de la premi√®re occurrence
langages = ["Python", "Java", "Python", "C++"]
print(f"\nIndex de 'Python': {langages.index('Python')}")
print(f"Index de 'Python' √† partir de l'index 2: {langages.index('Python', 2)}")

# count() - compte les occurrences
nombres = [1, 2, 2, 3, 2, 4, 2, 5]
print(f"\nNombre de 2 dans la liste: {nombres.count(2)}")
print(f"Nombre de 10 dans la liste: {nombres.count(10)}")
````

## 5. Op√©rateurs sur les listes

````python
# Op√©rateur + (concat√©nation)
liste1 = [1, 2, 3]
liste2 = [4, 5, 6]
liste3 = liste1 + liste2
print(f"liste1 + liste2 = {liste3}")

# Op√©rateur * (r√©p√©tition)
zeros = [0] * 5
print(f"\n[0] * 5 = {zeros}")

pattern = [1, 2] * 3
print(f"[1, 2] * 3 = {pattern}")

# Attention avec les listes imbriqu√©es!
# PI√àGE!
matrice_piege = [[0] * 3] * 2  # Toutes les lignes sont la M√äME liste!
print(f"\nMatrice pi√®ge: {matrice_piege}")
matrice_piege[0][0] = 1
print(f"Apr√®s modification: {matrice_piege}")  # Les deux lignes changent!

# Bonne m√©thode pour cr√©er une matrice
matrice_correcte = [[0] * 3 for _ in range(2)]
print(f"\nMatrice correcte: {matrice_correcte}")
matrice_correcte[0][0] = 1
print(f"Apr√®s modification: {matrice_correcte}")  # Seule la premi√®re ligne change

# Op√©rateur in et not in
nombres = [1, 2, 3, 4, 5]
print(f"\n3 in nombres: {3 in nombres}")
print(f"10 in nombres: {10 in nombres}")
print(f"10 not in nombres: {10 not in nombres}")
````

## 6. Listes imbriqu√©es (listes de listes)

````python
# Matrice 3x3
matrice = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Matrice:")
for ligne in matrice:
    print(ligne)

# Acc√®s aux √©l√©ments
print(f"\n√âl√©ment [0][0]: {matrice[0][0]}")
print(f"√âl√©ment [1][2]: {matrice[1][2]}")
print(f"√âl√©ment [2][2]: {matrice[2][2]}")

# Modification
matrice[1][1] = 50
print(f"\nApr√®s modification de [1][1]:")
for ligne in matrice:
    print(ligne)

# Parcours de matrice
print("\nParcours avec indices:")
for i in range(len(matrice)):
    for j in range(len(matrice[i])):
        print(f"matrice[{i}][{j}] = {matrice[i][j]}")

# Liste de listes h√©t√©rog√®nes
donnees = [
    ["Alice", 25, "Paris"],
    ["Bob", 30, "Lyon"],
    ["Charlie", 35, "Marseille"]
]

print("\nDonn√©es:")
for personne in donnees:
    nom, age, ville = personne
    print(f"{nom}, {age} ans, habite √† {ville}")
````

## 7. Copie shallow vs deep

**IMPORTANT** : Comprendre la diff√©rence entre copie shallow et deep est crucial pour √©viter des bugs.

````python
# Affectation simple (pas une copie!)
liste1 = [1, 2, 3]
liste2 = liste1  # liste2 pointe vers la M√äME liste

print(f"liste1: {liste1}")
print(f"liste2: {liste2}")
print(f"M√™me objet? {liste1 is liste2}")

liste2.append(4)
print(f"\nApr√®s liste2.append(4):")
print(f"liste1: {liste1}")  # liste1 est aussi modifi√©e!
print(f"liste2: {liste2}")

# Copie shallow (copie de surface)
liste1 = [1, 2, 3]
liste2 = liste1.copy()  # ou liste1[:] ou list(liste1)

print(f"\nApr√®s copie shallow:")
print(f"liste1: {liste1}")
print(f"liste2: {liste2}")
print(f"M√™me objet? {liste1 is liste2}")

liste2.append(4)
print(f"\nApr√®s liste2.append(4):")
print(f"liste1: {liste1}")  # liste1 non modifi√©e
print(f"liste2: {liste2}")

# PI√àGE avec shallow copy et listes imbriqu√©es
matrice1 = [[1, 2], [3, 4]]
matrice2 = matrice1.copy()  # Shallow copy

print(f"\nMatrice1: {matrice1}")
print(f"Matrice2: {matrice2}")

matrice2[0][0] = 99  # Modifie aussi matrice1!
print(f"\nApr√®s matrice2[0][0] = 99:")
print(f"Matrice1: {matrice1}")  # Modifi√©e!
print(f"Matrice2: {matrice2}")

# Copie deep (copie profonde)
import copy

matrice1 = [[1, 2], [3, 4]]
matrice2 = copy.deepcopy(matrice1)  # Deep copy

print(f"\nApr√®s deep copy:")
print(f"Matrice1: {matrice1}")
print(f"Matrice2: {matrice2}")

matrice2[0][0] = 99
print(f"\nApr√®s matrice2[0][0] = 99:")
print(f"Matrice1: {matrice1}")  # Non modifi√©e
print(f"Matrice2: {matrice2}")
````

## 8. It√©ration sur les listes

````python
langages = ["Python", "Java", "JavaScript", "C++"]

# It√©ration simple
print("It√©ration simple:")
for langage in langages:
    print(f"  {langage}")

# Avec index (enumerate)
print("\nAvec enumerate:")
for i, langage in enumerate(langages):
    print(f"  {i}: {langage}")

# enumerate avec start
print("\nAvec enumerate(start=1):")
for i, langage in enumerate(langages, start=1):
    print(f"  {i}. {langage}")

# It√©ration avec indices
print("\nAvec range(len(...)):")
for i in range(len(langages)):
    print(f"  langages[{i}] = {langages[i]}")

# It√©ration simultan√©e sur plusieurs listes (zip)
noms = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
villes = ["Paris", "Lyon", "Marseille"]

print("\nAvec zip:")
for nom, age, ville in zip(noms, ages, villes):
    print(f"  {nom}, {age} ans, {ville}")

# Compr√©hension de liste (list comprehension)
carres = [x**2 for x in range(1, 6)]
print(f"\nCarr√©s de 1 √† 5: {carres}")

# Avec condition
pairs = [x for x in range(1, 11) if x % 2 == 0]
print(f"Nombres pairs de 1 √† 10: {pairs}")

# Compr√©hension avec transformation
mots = ["python", "java", "c++"]
majuscules = [mot.upper() for mot in mots]
print(f"\nEn majuscules: {majuscules}")
````

## Pi√®ges courants

### 1. Modification pendant l'it√©ration

````python
# PI√àGE: Modifier une liste pendant qu'on l'it√®re
nombres = [1, 2, 3, 4, 5]

# MAUVAIS: comportement impr√©visible
print("MAUVAIS - modification pendant it√©ration:")
for nombre in nombres:
    if nombre % 2 == 0:
        nombres.remove(nombre)  # Danger!
print(f"R√©sultat: {nombres}")  # Peut manquer des √©l√©ments!

# SOLUTION 1: It√©rer sur une copie
nombres = [1, 2, 3, 4, 5]
print("\nSOLUTION 1 - it√©ration sur copie:")
for nombre in nombres.copy():
    if nombre % 2 == 0:
        nombres.remove(nombre)
print(f"R√©sultat: {nombres}")

# SOLUTION 2: Cr√©er une nouvelle liste
nombres = [1, 2, 3, 4, 5]
print("\nSOLUTION 2 - nouvelle liste:")
nombres_impairs = [n for n in nombres if n % 2 != 0]
print(f"R√©sultat: {nombres_impairs}")

# SOLUTION 3: It√©rer en sens inverse
nombres = [1, 2, 3, 4, 5]
print("\nSOLUTION 3 - it√©ration inverse:")
for i in range(len(nombres) - 1, -1, -1):
    if nombres[i] % 2 == 0:
        del nombres[i]
print(f"R√©sultat: {nombres}")
````

### 2. Shallow copy et r√©f√©rences partag√©es

````python
# PI√àGE: R√©f√©rences partag√©es
x = [1, 2, 3]
y = x
z = x

print(f"x: {x}")
print(f"y: {y}")
print(f"z: {z}")

y.append(4)
print(f"\nApr√®s y.append(4):")
print(f"x: {x}")  # Tous modifi√©s!
print(f"y: {y}")
print(f"z: {z}")

print(f"\nx is y: {x is y}")
print(f"x is z: {x is z}")

# SOLUTION: Cr√©er des copies
x = [1, 2, 3]
y = x.copy()
z = x[:]

y.append(4)
print(f"\nApr√®s copies et y.append(4):")
print(f"x: {x}")
print(f"y: {y}")
print(f"z: {z}")
````

### 3. Arguments par d√©faut mutables

````python
# PI√àGE: Liste comme argument par d√©faut
def ajouter_element_mauvais(element, liste=[]):
    liste.append(element)
    return liste

# La liste par d√©faut est cr√©√©e UNE SEULE FOIS!
print("MAUVAIS:")
print(ajouter_element_mauvais(1))  # [1]
print(ajouter_element_mauvais(2))  # [1, 2] - Pas ce qu'on veut!
print(ajouter_element_mauvais(3))  # [1, 2, 3] - Encore pire!

# SOLUTION: Utiliser None comme d√©faut
def ajouter_element_bon(element, liste=None):
    if liste is None:
        liste = []  # Nouvelle liste √† chaque appel
    liste.append(element)
    return liste

print("\nBON:")
print(ajouter_element_bon(1))  # [1]
print(ajouter_element_bon(2))  # [2]
print(ajouter_element_bon(3))  # [3]
````

### Solution Exercice 1

````python
def statistiques(nombres):
    if not nombres:
        return None
    
    nombres_tries = sorted(nombres)
    n = len(nombres_tries)
    
    # Calcul de la m√©diane
    if n % 2 == 0:
        mediane = (nombres_tries[n//2 - 1] + nombres_tries[n//2]) / 2
    else:
        mediane = nombres_tries[n//2]
    
    return {
        'somme': sum(nombres),
        'moyenne': sum(nombres) / len(nombres),
        'minimum': min(nombres),
        'maximum': max(nombres),
        'mediane': mediane
    }

# Tests
nombres = [5, 2, 8, 1, 9, 3, 7]
print("Statistiques:")
for cle, valeur in statistiques(nombres).items():
    print(f"  {cle}: {valeur}")

# Test avec liste vide
print(f"\nListe vide: {statistiques([])}")
````

````python
def transposer(matrice):
    if not matrice:
        return []
    
    # Solution 1: avec boucles
    nb_lignes = len(matrice)
    nb_colonnes = len(matrice[0])
    
    transposee = []
    for j in range(nb_colonnes):
        nouvelle_ligne = []
        for i in range(nb_lignes):
            nouvelle_ligne.append(matrice[i][j])
        transposee.append(nouvelle_ligne)
    
    return transposee

# Test
matrice = [
    [1, 2, 3],
    [4, 5, 6]
]

print("Matrice originale:")
for ligne in matrice:
    print(f"  {ligne}")

transposee = transposer(matrice)
print("\nMatrice transpos√©e:")
for ligne in transposee:
    print(f"  {ligne}")

# Solution 2: avec compr√©hension de liste
def transposer_v2(matrice):
    if not matrice:
        return []
    return [[matrice[i][j] for i in range(len(matrice))] 
            for j in range(len(matrice[0]))]

print("\nAvec compr√©hension:")
for ligne in transposer_v2(matrice):
    print(f"  {ligne}")

# Solution 3: avec zip (la plus pythonique)
def transposer_v3(matrice):
    return [list(colonne) for colonne in zip(*matrice)]

print("\nAvec zip:")
for ligne in transposer_v3(matrice):
    print(f"  {ligne}")
````
