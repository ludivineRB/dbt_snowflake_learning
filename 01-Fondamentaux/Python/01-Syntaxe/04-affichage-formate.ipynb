## 2. F-strings (Python 3.6+)

Les **f-strings** (formatted string literals) sont la m√©thode moderne et recommand√©e pour le formatage.

### Syntaxe de base

````python
# Pr√©fixer la cha√Æne avec 'f' et utiliser {}
nom = "Alice"
age = 25

message = f"Bonjour, je m'appelle {nom} et j'ai {age} ans."
print(message)

# Expressions directement dans les accolades
print(f"Dans 5 ans, j'aurai {age + 5} ans.")
print(f"Le double de {age} est {age * 2}.")
````

### Expressions complexes

````python
# Appels de fonctions
texte = "python"
print(f"Majuscules : {texte.upper()}")
print(f"Longueur : {len(texte)}")

# Op√©rations
a, b = 10, 3
print(f"{a} + {b} = {a + b}")
print(f"{a} / {b} = {a / b}")

# Conditions ternaires
age = 18
print(f"Statut : {'Majeur' if age >= 18 else 'Mineur'}")

# Acc√®s √† des attributs
import math
print(f"Pi : {math.pi}")
````

### Formatage des nombres

#### D√©cimales

````python
# Contr√¥ler le nombre de d√©cimales : {valeur:.Nf}
pi = 3.14159265359

print(f"Pi : {pi}")
print(f"Pi (2 d√©cimales) : {pi:.2f}")
print(f"Pi (4 d√©cimales) : {pi:.4f}")
print(f"Pi (0 d√©cimale) : {pi:.0f}")

# Nombres tr√®s grands
nombre = 1234567.89
print(f"\nSans formatage : {nombre}")
print(f"2 d√©cimales : {nombre:.2f}")
print(f"Avec s√©parateurs : {nombre:,.2f}")
````

#### Notation scientifique

````python
# Notation scientifique : {valeur:e}
grand = 1234567890
petit = 0.000000123

print(f"Standard : {grand}")
print(f"Scientifique : {grand:e}")
print(f"Scientifique (2 d√©c.) : {grand:.2e}")

print(f"\nPetit nombre : {petit}")
print(f"Scientifique : {petit:e}")
````

#### Pourcentages

````python
# Formatage en pourcentage : {valeur:%}
taux = 0.1525

print(f"Taux : {taux}")
print(f"Taux : {taux:%}")
print(f"Taux : {taux:.2%}")
print(f"Taux : {taux:.1%}")

# Application pratique
reussite = 45 / 50
print(f"\nTaux de r√©ussite : {reussite:.1%}")
````

### Alignement et padding

#### Largeur minimale

````python
# Largeur minimale : {valeur:N}
nombre = 42
print(f"Sans padding : [{nombre}]")
print(f"Largeur 10 : [{nombre:10}]")
print(f"Largeur 10 (float) : [{nombre:10.2f}]")
````

#### Alignement

````python
# < : gauche (d√©faut pour str)
# > : droite (d√©faut pour nombres)
# ^ : centre

texte = "Python"
print(f"Gauche : [{texte:<15}]")
print(f"Droite : [{texte:>15}]")
print(f"Centre : [{texte:^15}]")

# Avec des nombres
nombre = 42
print(f"\nNombre gauche : [{nombre:<10}]")
print(f"Nombre droite : [{nombre:>10}]")
print(f"Nombre centre : [{nombre:^10}]")
````

#### Caract√®re de remplissage

````python
# Sp√©cifier le caract√®re de remplissage : {valeur:C<N}
nombre = 42

print(f"Avec des 0 : {nombre:0>10}")
print(f"Avec des * : {nombre:*>10}")
print(f"Avec des - : {nombre:-^10}")

# Application : num√©ros format√©s
numero = 7
print(f"\nNum√©ro : {numero:0>4}")  # 0007

# Titre centr√©
titre = "Menu Principal"
print(f"{titre:=^50}")
````

### Formatage combin√©

````python
# Combiner alignement, largeur et pr√©cision
prix = 19.99
quantite = 3

print(f"{'Article':<20} {'Quantit√©':>10} {'Prix':>10}")
print("-" * 45)
print(f"{'Pommes':<20} {quantite:>10} {prix:>10.2f} ‚Ç¨")
print(f"{'Oranges':<20} {5:>10} {12.50:>10.2f} ‚Ç¨")
print(f"{'Bananes':<20} {8:>10} {8.75:>10.2f} ‚Ç¨")
print("-" * 45)
print(f"{'TOTAL':<20} {16:>10} {41.24:>10.2f} ‚Ç¨")
````

### Formats sp√©ciaux

#### Binaire, octal, hexad√©cimal

````python
# b : binaire, o : octal, x : hexad√©cimal (minuscules), X : hexad√©cimal (majuscules)
nombre = 42

print(f"D√©cimal : {nombre}")
print(f"Binaire : {nombre:b}")
print(f"Binaire (pr√©fixe) : {nombre:#b}")
print(f"Octal : {nombre:o}")
print(f"Octal (pr√©fixe) : {nombre:#o}")
print(f"Hexad√©cimal : {nombre:x}")
print(f"Hexad√©cimal (pr√©fixe) : {nombre:#x}")
print(f"Hexad√©cimal (maj.) : {nombre:#X}")
````

#### S√©parateurs de milliers

````python
# , : virgule comme s√©parateur de milliers
# _ : underscore comme s√©parateur de milliers
population = 67000000
prix = 1234567.89

print(f"Population : {population}")
print(f"Population : {population:,}")
print(f"Population : {population:_}")

print(f"\nPrix : {prix:.2f}")
print(f"Prix : {prix:,.2f}")
print(f"Prix : {prix:_.2f}")
````

### Accolades litt√©rales

````python
# Pour afficher des accolades litt√©rales, les doubler : {{ et }}
nom = "Python"
print(f"{{nom}} devient {nom}")
print(f"Voici un dictionnaire : {{cl√©: {nom}}}")

# Exemple : g√©n√©ration de code
variable = "x"
valeur = 10
print(f"let {variable} = {valeur};")
print(f"const obj = {{ {variable}: {valeur} }};")
````

## 3. M√©thode str.format()

Avant les f-strings, `str.format()` √©tait la m√©thode recommand√©e.

### Syntaxe de base

````python
# Placeholders {} remplac√©s par les arguments de format()
message = "Bonjour, {} !".format("Alice")
print(message)

# Plusieurs arguments
message = "Je m'appelle {} et j'ai {} ans.".format("Bob", 30)
print(message)

# Index explicites
message = "{0} + {1} = {2}".format(5, 3, 8)
print(message)

# R√©utiliser un argument
message = "{0} x {0} = {1}".format(5, 25)
print(message)
````

### Arguments nomm√©s

````python
# Utiliser des noms au lieu d'indices
message = "Bonjour, {prenom} {nom} !".format(prenom="Alice", nom="Dupont")
print(message)

# M√©lange index et noms
message = "{0} a {age} ans et habite √† {ville}.".format(
    "Bob", age=25, ville="Paris"
)
print(message)
````

### Formatage avec format()

````python
# M√™me syntaxe de formatage que les f-strings
pi = 3.14159

print("Pi : {:.2f}".format(pi))
print("Pi : {:.4f}".format(pi))

# Avec noms
print("Prix : {prix:.2f} ‚Ç¨".format(prix=19.99))

# Alignement
print("{:<10} {:>10}".format("Gauche", "Droite"))
print("{:<10} {:>10}".format("Left", "Right"))
````

## 4. Op√©rateur % (Ancien Style)

L'op√©rateur `%` est l'ancienne m√©thode de formatage (h√©rit√©e du C).

**Note :** Bien que toujours support√©, il est recommand√© d'utiliser les f-strings.

````python
# %s : cha√Æne, %d : entier, %f : flottant
nom = "Alice"
age = 25

message = "Je m'appelle %s et j'ai %d ans." % (nom, age)
print(message)

# Formatage de nombres
pi = 3.14159
print("Pi : %.2f" % pi)
print("Pi : %.4f" % pi)

# Avec un dictionnaire
message = "%(prenom)s %(nom)s" % {"prenom": "Bob", "nom": "Martin"}
print(message)
````

### Comparaison des 3 m√©thodes

````python
# M√™me r√©sultat avec les 3 m√©thodes
nom = "Alice"
age = 25

# 1. F-string (recommand√©)
print(f"Je m'appelle {nom} et j'ai {age} ans.")

# 2. str.format()
print("Je m'appelle {} et j'ai {} ans.".format(nom, age))

# 3. Op√©rateur %
print("Je m'appelle %s et j'ai %d ans." % (nom, age))
````

## 5. repr() vs str()

### Diff√©rences

````python
# str() : repr√©sentation lisible pour l'utilisateur
# repr() : repr√©sentation technique pour les d√©veloppeurs

texte = "Hello\nWorld"

print("str() :", str(texte))
print("repr() :", repr(texte))

# Avec les nombres
from datetime import datetime
maintenant = datetime.now()

print("\nstr() :", str(maintenant))
print("repr() :", repr(maintenant))
````

````python
# Dans les f-strings : !s (str), !r (repr), !a (ascii)
texte = "Caf√© ‚òï"

print(f"D√©faut : {texte}")
print(f"str : {texte!s}")
print(f"repr : {texte!r}")
print(f"ascii : {texte!a}")
````

### Pi√®ge 2 : print() vs return

````python
# ‚ùå MAUVAIS : Confusion entre print() et return
def calculer_somme_mauvais(a, b):
    print(a + b)  # Affiche, mais ne retourne rien

resultat = calculer_somme_mauvais(5, 3)
print(f"R√©sultat : {resultat}")  # None

# ‚úÖ BON : Utiliser return
def calculer_somme_bon(a, b):
    return a + b

resultat = calculer_somme_bon(5, 3)
print(f"R√©sultat : {resultat}")  # 8
````

### Pi√®ge 4 : Formatage et pr√©cision

## Solutions

### Solution Exercice 1

````python
# Donn√©es
produits = [
    ("Pommes", 3, 1.50),
    ("Pain", 2, 2.30),
    ("Lait", 1, 1.20)
]

# En-t√™te
print("=" * 70)
print(f"{'TICKET DE CAISSE':^70}")
print("=" * 70)
print(f"{'Article':<20} {'Qt√©':>10} {'P.U.':>15} {'Total':>15}")
print("-" * 70)

# Lignes
total_general = 0
for article, quantite, prix_unitaire in produits:
    total_ligne = quantite * prix_unitaire
    total_general += total_ligne
    print(f"{article:<20} {quantite:>10} {prix_unitaire:>15.2f} ‚Ç¨ {total_ligne:>15.2f} ‚Ç¨")

# Total
print("-" * 70)
print(f"{'TOTAL':<20} {'':>10} {'':>15} {total_general:>15.2f} ‚Ç¨")
print("=" * 70)
print(f"Merci de votre visite !")
print("=" * 70)
````

### Solution Exercice 2

````python
# Donn√©es
employes = [
    ("Alice", 25, "Paris", 35000.50),
    ("Bob", 30, "Lyon", 42000.00),
    ("Charlie", 35, "Marseille", 38500.75)
]

# En-t√™te
print(f"{'Nom':<15} {'Age':>5} {'Ville':<15} {'Salaire':>12}")
print("-" * 50)

# Lignes
for nom, age, ville, salaire in employes:
    print(f"{nom:<15} {age:>5} {ville:<15} {salaire:>12,.2f} ‚Ç¨")

# Statistiques
print("-" * 50)
total_salaires = sum(emp[3] for emp in employes)
moyenne_salaires = total_salaires / len(employes)
print(f"{'TOTAL':<15} {'':>5} {'':>15} {total_salaires:>12,.2f} ‚Ç¨")
print(f"{'MOYENNE':<15} {'':>5} {'':>15} {moyenne_salaires:>12,.2f} ‚Ç¨")
````

### Solution Exercice 3

````python
def convertir_temperature(celsius):
    """
    Convertit une temp√©rature de Celsius vers Fahrenheit et Kelvin.
    
    Args:
        celsius (float): Temp√©rature en degr√©s Celsius
    """
    # Conversions
    fahrenheit = celsius * 9/5 + 32
    kelvin = celsius + 273.15
    
    # Affichage
    print("=" * 40)
    print(f"{'CONVERTISSEUR DE TEMP√âRATURE':^40}")
    print("=" * 40)
    print(f"Celsius    : {celsius:>10.1f} ¬∞C")
    print(f"Fahrenheit : {fahrenheit:>10.1f} ¬∞F")
    print(f"Kelvin     : {kelvin:>10.1f} K")
    print("=" * 40)

# Tests
convertir_temperature(0)     # Point de cong√©lation de l'eau
print()
convertir_temperature(100)   # Point d'√©bullition de l'eau
print()
convertir_temperature(25)    # Temp√©rature ambiante
print()
convertir_temperature(-40)   # -40¬∞C = -40¬∞F !

# Avec saisie utilisateur (d√©commentez pour tester)
# celsius = float(input("Entrez une temp√©rature en Celsius : "))
# convertir_temperature(celsius)
````
