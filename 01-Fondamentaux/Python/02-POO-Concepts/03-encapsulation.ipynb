### 1.4 Pourquoi l'encapsulation ?

### 1.5 Citation c√©l√®bre

## 2. Niveaux de Visibilit√©

### 2.1 Les trois niveaux (th√©oriques)

En th√©orie de la POO, il existe trois niveaux de visibilit√© :

#### Public (+)
- **Accessible de partout**
- Fait partie de l'interface publique de la classe
- Peut √™tre utilis√© par tout le monde
- **Engagement** : Ne pas changer sans r√©fl√©chir aux cons√©quences

#### Prot√©g√© (#)
- **Accessible dans la classe et ses sous-classes**
- Utilis√© pour l'h√©ritage
- Pas accessible de l'ext√©rieur

#### Priv√© (-)
- **Accessible uniquement √† l'int√©rieur de la classe**
- D√©tails d'impl√©mentation internes
- Peut √™tre modifi√© librement sans impact externe

### 2.2 Impl√©mentation en Python

**Important** : Python n'a pas de v√©ritable encapsulation forte (pas de mot-cl√© `private`).

Python utilise des **conventions de nommage** :

| Convention | Visibilit√© | Exemple | Signification |

| `nom` | Public | `self.nom` | Accessible partout |
| `_nom` | Prot√©g√© | `self._solde` | "Svp ne touchez pas, mais c'est possible" |
| `__nom` | Priv√© | `self.__secret` | Name mangling (transformation du nom) |

### 2.3 Name Mangling

Avec `__attribut`, Python renomme automatiquement l'attribut en `_NomClasse__attribut`.

**But** : √âviter les conflits de noms dans l'h√©ritage, pas une vraie protection.

## 3. Getters et Setters

### 3.1 Qu'est-ce qu'un getter/setter ?

- **Getter** : M√©thode pour **lire** (get) la valeur d'un attribut priv√©
- **Setter** : M√©thode pour **modifier** (set) la valeur d'un attribut priv√©

### 3.2 Pourquoi utiliser des getters/setters ?

#### 1. Validation
```python

# Avec setter : validation

```

#### 2. Contr√¥le d'acc√®s
```python
# Lecture seule

    def get_age(self):  # Getter public
        return self._age
    # Pas de setter ‚Üí l'√¢ge ne peut pas √™tre modifi√© de l'ext√©rieur
```

#### 3. Calcul √† la vol√©e
```python
class Rectangle:
    def get_aire(self):  # Calcul√©, pas stock√©
        return self._largeur * self._hauteur
```

#### 4. Logging et d√©bogage
```python
def set_prix(self, prix):
    print(f"Prix chang√© de {self._prix} √† {prix}")  # Log
    self._prix = prix
```

#### 5. √âvolution du code
```python

# Version 2 : besoin de convertir Celsius ‚Üí Fahrenheit
# Avec getter/setter, on peut ajouter la conversion sans changer l'interface
```

### 3.3 Le pattern classique (style Java)

```python

        self._nom = nom  # Attribut priv√©
```

**Probl√®me** : Verbeux et peu pythonique.

### 3.4 Le style Python : @property

Python offre le d√©corateur `@property` pour cr√©er des getters/setters √©l√©gants :


### 3.5 Quand NE PAS utiliser de getters/setters ?

**Anti-pattern** : Getters/setters inutiles


**R√®gle** : Commencez simple (attributs publics). Ajoutez des getters/setters seulement quand vous en avez **besoin**.

## 4. Invariants de Classe

### 4.1 D√©finition

Un **invariant de classe** est une **propri√©t√© qui doit toujours √™tre vraie** pour un objet, quel que soit son √©tat.

### 4.2 Exemples d'invariants

#### CompteBancaire
- Invariant : `solde >= 0` (pas de d√©couvert)
- Toute op√©ration doit maintenir cette propri√©t√©

#### Rectangle
- Invariants : `largeur > 0` et `hauteur > 0`
- Un rectangle ne peut pas avoir de dimensions n√©gatives

#### Date
- Invariants : `1 <= jour <= 31`, `1 <= mois <= 12`, `annee > 0`
- Une date doit toujours √™tre valide

### 4.3 R√¥le de l'encapsulation

L'encapsulation permet de **garantir les invariants** en :
1. Cachant les attributs (pas d'acc√®s direct)
2. Validant dans les setters
3. S'assurant que toutes les m√©thodes maintiennent les invariants

### 4.4 Exemple

```python
class CompteBancaire:
    def __init__(self, solde_initial):
        if solde_initial < 0:
            raise ValueError("Solde initial ne peut √™tre n√©gatif")
        self._solde = solde_initial  # INVARIANT: _solde >= 0
    
    def retirer(self, montant):
        if self._solde - montant < 0:  # V√âRIFICATION de l'invariant
            raise ValueError("Solde insuffisant")
        self._solde -= montant
        # INVARIANT maintenu : _solde >= 0
```

Sans encapsulation :
```python

```

### 4.5 Avantages

- **Fiabilit√©** : L'objet est toujours dans un √©tat coh√©rent
- **Debugging** : Si un bug appara√Æt, il ne peut venir que des m√©thodes de la classe
- **Raisonnement** : On peut faire des hypoth√®ses sur l'√©tat de l'objet

## 5. Couplage Faible vs Couplage Fort

### 5.1 D√©finitions

#### Couplage Fort (tight coupling)
- D√©pendance forte entre les classes
- Conna√Ætre les d√©tails d'impl√©mentation
- Difficile √† modifier, √† tester, √† r√©utiliser

#### Couplage Faible (loose coupling)
- D√©pendance minimale
- Interaction via des interfaces bien d√©finies
- Facile √† modifier, √† tester, √† r√©utiliser

### 5.2 Exemple de couplage fort

```python
# ‚ùå COUPLAGE FORT
class Voiture:
    def __init__(self):
        self.essence = 50  # Public

class Station:
    def faire_le_plein(self, voiture):
        voiture.essence = 100  # Acc√®s direct √† l'attribut
        # Si Voiture change 'essence' en 'carburant', Station casse !
```

### 5.3 Exemple de couplage faible

```python
# ‚úÖ COUPLAGE FAIBLE
class Voiture:
    def __init__(self):
        self._essence = 50  # Priv√©
    
    def remplir_reservoir(self):
        self._essence = 100  # Interface publique

    def faire_le_plein(self, voiture):
        voiture.remplir_reservoir()  # Utilise l'interface
        # L'impl√©mentation interne de Voiture peut changer librement
```

### 5.4 Avantages du couplage faible

1. **Maintenance** : Changer une classe n'affecte pas les autres
2. **Tests** : Facile de cr√©er des mocks/stubs
3. **R√©utilisation** : Les classes sont plus ind√©pendantes
4. **√âvolution** : On peut remplacer une impl√©mentation par une autre

### 5.5 Comment obtenir un couplage faible ?

- ‚úÖ Utiliser l'encapsulation (cacher les d√©tails internes)
- ‚úÖ D√©finir des interfaces claires
- ‚úÖ D√©pendre des abstractions, pas des impl√©mentations
- ‚úÖ Principe de responsabilit√© unique (SRP)
- ‚ùå √âviter les acc√®s directs aux attributs d'autres classes

````python
# Exemple 1 : Niveaux de visibilit√© en Python

class Personne:
    """D√©monstration des niveaux de visibilit√©"""
    
    def __init__(self, nom, age, numero_secu):
        # PUBLIC : accessible partout
        self.nom = nom
        
        # PROT√âG√â : convention "ne touchez pas, mais c'est possible"
        self._age = age
        
        # PRIV√â : name mangling appliqu√©
        self.__numero_secu = numero_secu
    
    def afficher_info(self):
        """M√©thode publique utilisant tous les attributs"""
        print(f"Nom: {self.nom}")
        print(f"√Çge: {self._age}")
        print(f"Num√©ro s√©curit√© sociale: {self.__numero_secu}")
    
    def _methode_protegee(self):
        """M√©thode prot√©g√©e (convention)"""
        return "M√©thode interne, ne pas appeler de l'ext√©rieur"
    
    def __methode_privee(self):
        """M√©thode priv√©e (name mangling)"""
        return "M√©thode vraiment priv√©e"

print("=== Exemple 1 : Visibilit√© ===")

p = Personne("Alice Dupont", 30, "123456789012345")

# ‚úÖ Public : acc√®s direct OK
print(f"\nPublic - Nom: {p.nom}")
p.nom = "Alice Martin"  # Modification OK
print(f"Nom modifi√©: {p.nom}")

# ‚ö†Ô∏è Prot√©g√© : acc√®s possible mais d√©conseill√©
print(f"\nProt√©g√© - √Çge: {p._age}")
print("(On peut y acc√©der, mais c'est d√©conseill√© par convention)")

# ‚ùå Priv√© : acc√®s direct impossible
try:
    print(p.__numero_secu)
except AttributeError as e:
    print(f"\nPriv√© - Erreur: {e}")
    print("L'attribut __numero_secu n'est pas accessible directement")

# Mais on peut contourner avec name mangling (pas recommand√© !)
print(f"\nContournement (name mangling): {p._Personne__numero_secu}")
print("(Possible mais tr√®s d√©conseill√© !)")

# ‚úÖ Utilisation correcte : via une m√©thode publique
print("\nUtilisation correcte via m√©thode publique :")
p.afficher_info()
````

````python
# Exemple 2 : Getters et setters avec @property

class CompteBancaire:
    """Compte bancaire avec encapsulation et validation"""
    
    def __init__(self, numero, titulaire, solde_initial=0):
        self._numero = numero  # Lecture seule
        self._titulaire = titulaire  # Modifiable avec validation
        self._solde = 0
        self.solde = solde_initial  # Utilise le setter pour validation
    
    # GETTER : Lecture seule du num√©ro de compte
    @property
    def numero(self):
        """Num√©ro de compte (lecture seule)"""
        return self._numero
    # Pas de setter ‚Üí Le num√©ro ne peut pas √™tre modifi√©
    
    # GETTER + SETTER : Titulaire avec validation
    @property
    def titulaire(self):
        """Nom du titulaire"""
        return self._titulaire
    
    @titulaire.setter
    def titulaire(self, nom):
        """Modifie le titulaire avec validation"""
        if not nom or not isinstance(nom, str):
            raise ValueError("Le nom du titulaire doit √™tre une cha√Æne non vide")
        if len(nom) < 2:
            raise ValueError("Le nom doit contenir au moins 2 caract√®res")
        self._titulaire = nom.strip()
        print(f"Titulaire mis √† jour: {self._titulaire}")
    
    # GETTER + SETTER : Solde avec validation forte
    @property
    def solde(self):
        """Solde du compte"""
        return self._solde
    
    @solde.setter
    def solde(self, montant):
        """Modifie le solde avec validation (INVARIANT: solde >= 0)"""
        if montant < 0:
            raise ValueError("Le solde ne peut pas √™tre n√©gatif")
        self._solde = montant
    
    # PROPRI√âT√â CALCUL√âE (getter sans setter)
    @property
    def solde_formate(self):
        """Solde format√© avec devise (calcul√© √† la vol√©e)"""
        return f"{self._solde:.2f} ‚Ç¨"
    
    def deposer(self, montant):
        """D√©pose de l'argent"""
        if montant <= 0:
            raise ValueError("Le montant doit √™tre positif")
        self._solde += montant
        print(f"D√©p√¥t de {montant}‚Ç¨. Nouveau solde: {self.solde_formate}")
    
    def retirer(self, montant):
        """Retire de l'argent (respecte l'invariant)"""
        if montant <= 0:
            raise ValueError("Le montant doit √™tre positif")
        if montant > self._solde:
            raise ValueError(f"Solde insuffisant. Disponible: {self.solde_formate}")
        self._solde -= montant
        print(f"Retrait de {montant}‚Ç¨. Nouveau solde: {self.solde_formate}")

print("\n=== Exemple 2 : @property ===")

compte = CompteBancaire("FR123456", "Alice Dupont", 1000)

# ‚úÖ Lecture via getter (syntaxe d'attribut)
print(f"\nNum√©ro: {compte.numero}")
print(f"Titulaire: {compte.titulaire}")
print(f"Solde: {compte.solde_formate}")

# ‚úÖ Modification via setter (syntaxe d'attribut)
print("\nModification du titulaire :")
compte.titulaire = "  Alice Martin  "  # Appelle le setter avec validation

# ‚ùå Tentative de modification du num√©ro (lecture seule)
print("\nTentative de modification du num√©ro (lecture seule) :")
try:
    compte.numero = "FR999999"
except AttributeError as e:
    print(f"‚úó Erreur: {e}")

# ‚ùå Tentative de solde n√©gatif
print("\nTentative de solde n√©gatif :")
try:
    compte.solde = -500
except ValueError as e:
    print(f"‚úó Erreur: {e}")

# ‚úÖ Op√©rations respectant les invariants
print("\nOp√©rations valides :")
compte.deposer(500)
compte.retirer(200)
````

````python
# Exemple 3 : Invariants de classe

class Rectangle:
    """Rectangle avec invariants : largeur > 0 et hauteur > 0"""
    
    def __init__(self, largeur, hauteur):
        """Initialise un rectangle avec validation des invariants"""
        self.largeur = largeur  # Utilise le setter
        self.hauteur = hauteur  # Utilise le setter
    
    @property
    def largeur(self):
        return self._largeur
    
    @largeur.setter
    def largeur(self, valeur):
        """INVARIANT: largeur > 0"""
        if valeur <= 0:
            raise ValueError("La largeur doit √™tre strictement positive")
        self._largeur = valeur
    
    @property
    def hauteur(self):
        return self._hauteur
    
    @hauteur.setter
    def hauteur(self, valeur):
        """INVARIANT: hauteur > 0"""
        if valeur <= 0:
            raise ValueError("La hauteur doit √™tre strictement positive")
        self._hauteur = valeur
    
    @property
    def aire(self):
        """Aire calcul√©e √† la vol√©e"""
        return self._largeur * self._hauteur
    
    @property
    def perimetre(self):
        """P√©rim√®tre calcul√© √† la vol√©e"""
        return 2 * (self._largeur + self._hauteur)
    
    def redimensionner(self, facteur):
        """Redimensionne le rectangle (respecte les invariants)"""
        if facteur <= 0:
            raise ValueError("Le facteur doit √™tre positif")
        # Les setters valident automatiquement les nouveaux valeurs
        self.largeur = self._largeur * facteur
        self.hauteur = self._hauteur * facteur
        print(f"Rectangle redimensionn√© (x{facteur}): {self._largeur}x{self._hauteur}")
    
    def __repr__(self):
        return f"Rectangle({self._largeur}, {self._hauteur})"

print("\n=== Exemple 3 : Invariants ===")

# ‚úÖ Cr√©ation valide
rect = Rectangle(10, 5)
print(f"\nRectangle cr√©√©: {rect}")
print(f"Aire: {rect.aire}")
print(f"P√©rim√®tre: {rect.perimetre}")

# ‚úÖ Modification valide
print("\nModification valide :")
rect.largeur = 15
print(f"Nouvelle largeur: {rect.largeur}")
print(f"Nouvelle aire: {rect.aire}")

# ‚ùå Tentative de violation d'invariant
print("\nTentative de largeur n√©gative :")
try:
    rect.largeur = -5
except ValueError as e:
    print(f"‚úó Erreur: {e}")
    print("‚Üí L'invariant est prot√©g√© !")

print("\nTentative de hauteur nulle :")
try:
    rect.hauteur = 0
except ValueError as e:
    print(f"‚úó Erreur: {e}")

# ‚ùå Cr√©ation invalide
print("\nTentative de cr√©ation avec dimensions invalides :")
try:
    rect_invalide = Rectangle(10, -5)
except ValueError as e:
    print(f"‚úó Erreur: {e}")

# ‚úÖ Redimensionnement (respecte les invariants)
print("\nRedimensionnement :")
rect.redimensionner(0.5)
print(f"Nouvelle aire: {rect.aire}")
````

````python
# Exemple 4 : Couplage faible vs couplage fort

print("\n=== Exemple 4 : Couplage ===")

# ‚ùå VERSION COUPLAGE FORT
print("\n--- Couplage Fort (mauvais) ---")

class MoteurFort:
    def __init__(self):
        self.etat = "arr√™t√©"  # Attribut public
        self.carburant = 50   # Attribut public

class VoitureFort:
    def __init__(self):
        self.moteur = MoteurFort()
    
    def demarrer(self):
        # Acc√®s direct aux attributs du moteur (couplage fort)
        if self.moteur.carburant > 0:
            self.moteur.etat = "d√©marr√©"  # Modification directe
            print(f"Moteur {self.moteur.etat}")
        else:
            print("Pas de carburant")

voiture1 = VoitureFort()
voiture1.demarrer()

# Probl√®me : Si MoteurFort change 'etat' en 'status', VoitureFort casse !
print("‚Üí Probl√®me : VoitureFort d√©pend des d√©tails internes de MoteurFort")

# ‚úÖ VERSION COUPLAGE FAIBLE
print("\n--- Couplage Faible (bon) ---")

class MoteurFaible:
    def __init__(self):
        self._etat = "arr√™t√©"  # Priv√©
        self._carburant = 50   # Priv√©
    
    def demarrer(self):  # Interface publique
        """D√©marre le moteur si possible"""
        if self._carburant > 0:
            self._etat = "d√©marr√©"
            return True
        return False
    
    def arreter(self):  # Interface publique
        """Arr√™te le moteur"""
        self._etat = "arr√™t√©"
    
    def est_demarre(self):  # Interface publique
        """V√©rifie si le moteur est d√©marr√©"""
        return self._etat == "d√©marr√©"

class VoitureFaible:
    def __init__(self):
        self._moteur = MoteurFaible()
    
    def demarrer(self):
        # Utilise l'interface publique (couplage faible)
        if self._moteur.demarrer():
            print("Voiture d√©marr√©e")
        else:
            print("Impossible de d√©marrer (pas de carburant)")
    
    def arreter(self):
        self._moteur.arreter()
        print("Voiture arr√™t√©e")

voiture2 = VoitureFaible()
voiture2.demarrer()
voiture2.arreter()

print("\n‚Üí Avantage : VoitureFaible utilise l'interface de MoteurFaible")
print("  L'impl√©mentation interne de MoteurFaible peut changer librement")
print("  sans affecter VoitureFaible")
````

````python
# Exemple 5 : Encapsulation compl√®te - Classe Temperature

class Temperature:
    """G√®re une temp√©rature avec conversion Celsius/Fahrenheit"""
    
    def __init__(self, celsius=0):
        """Initialise avec une temp√©rature en Celsius"""
        self._celsius = celsius  # Stockage interne en Celsius
    
    @property
    def celsius(self):
        """Temp√©rature en degr√©s Celsius"""
        return self._celsius
    
    @celsius.setter
    def celsius(self, valeur):
        """D√©finit la temp√©rature en Celsius (validation)"""
        if valeur < -273.15:  # Z√©ro absolu
            raise ValueError("Temp√©rature en dessous du z√©ro absolu (-273.15¬∞C)")
        self._celsius = valeur
    
    @property
    def fahrenheit(self):
        """Temp√©rature en degr√©s Fahrenheit (calcul√©)"""
        return self._celsius * 9/5 + 32
    
    @fahrenheit.setter
    def fahrenheit(self, valeur):
        """D√©finit la temp√©rature en Fahrenheit (converti en Celsius)"""
        celsius_equivalent = (valeur - 32) * 5/9
        self.celsius = celsius_equivalent  # Utilise le setter Celsius (validation)
    
    @property
    def kelvin(self):
        """Temp√©rature en Kelvin (calcul√©)"""
        return self._celsius + 273.15
    
    @kelvin.setter
    def kelvin(self, valeur):
        """D√©finit la temp√©rature en Kelvin"""
        self.celsius = valeur - 273.15
    
    def __repr__(self):
        return f"Temperature({self._celsius}¬∞C)"
    
    def __str__(self):
        return f"{self._celsius:.1f}¬∞C / {self.fahrenheit:.1f}¬∞F / {self.kelvin:.1f}K"

print("\n=== Exemple 5 : Encapsulation et conversion ===")

temp = Temperature(25)
print(f"\nTemp√©rature initiale: {temp}")

print("\nLecture dans diff√©rentes unit√©s :")
print(f"  Celsius: {temp.celsius}¬∞C")
print(f"  Fahrenheit: {temp.fahrenheit:.1f}¬∞F")
print(f"  Kelvin: {temp.kelvin:.1f}K")

print("\nModification en Fahrenheit :")
temp.fahrenheit = 32  # 0¬∞C
print(f"  Nouvelle temp√©rature: {temp}")

print("\nModification en Kelvin :")
temp.kelvin = 300
print(f"  Nouvelle temp√©rature: {temp}")

print("\nTentative de temp√©rature impossible :")
try:
    temp.celsius = -300  # En dessous du z√©ro absolu
except ValueError as e:
    print(f"  ‚úó Erreur: {e}")

print("\n‚Üí Avantages de cette encapsulation :")
print("  1. Une seule valeur stock√©e (_celsius)")
print("  2. Conversions automatiques transparentes")
print("  3. Validation centralis√©e (z√©ro absolu)")
print("  4. Interface simple et intuitive")
````

## 7. Pi√®ges Courants

### Pi√®ge 1 : Tout mettre en public

**Erreur** : Ne jamais utiliser `_attribut` par paresse.

**Cons√©quence** :
- Pas de protection des invariants
- Modification directe possible de l'ext√©rieur
- Code difficile √† maintenir

**Solution** : Commencer par des attributs priv√©s/prot√©g√©s, exposer via des propri√©t√©s si n√©cessaire.

### Pi√®ge 2 : Getters/setters inutiles (Anemic Domain Model)

**Anti-pattern** : Classe avec que des getters/setters, aucune logique m√©tier.

````python
# D√©monstration du pi√®ge 2

# ‚ùå Anemic Domain Model (mauvais)
class PersonneAnemic:
    def __init__(self):
        self._nom = ""
        self._age = 0
    
    @property
    def nom(self):
        return self._nom
    
    @nom.setter
    def nom(self, nom):
        self._nom = nom  # Aucune logique
    
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, age):
        self._age = age  # Aucune logique

# Probl√®me : C'est juste un conteneur de donn√©es, pas un vrai objet
# Un dictionnaire ferait le m√™me travail !

# ‚úÖ Rich Domain Model (bon)
class PersonneRiche:
    def __init__(self, nom, annee_naissance):
        self.nom = nom  # Public si pas de validation n√©cessaire
        self._annee_naissance = annee_naissance
    
    @property
    def age(self):
        """Calcul√© √† la vol√©e (logique m√©tier)"""
        from datetime import datetime
        return datetime.now().year - self._annee_naissance
    
    def est_majeur(self):
        """Logique m√©tier"""
        return self.age >= 18
    
    def peut_voter(self):
        """Logique m√©tier"""
        return self.age >= 18
    
    def peut_conduire(self):
        """Logique m√©tier"""
        return self.age >= 18

print("\n=== Pi√®ge 2 : Anemic vs Rich Domain Model ===")
p = PersonneRiche("Alice", 1990)
print(f"\n{p.nom} a {p.age} ans")
print(f"Majeur ? {p.est_majeur()}")
print(f"Peut voter ? {p.peut_voter()}")

print("\n‚Üí La classe contient de la LOGIQUE M√âTIER, pas juste des donn√©es")
````

### Pi√®ge 3 : Modification d'objets mutables retourn√©s

**Erreur** : Retourner une r√©f√©rence √† un objet mutable interne.

````python
# D√©monstration du pi√®ge 3

# ‚ùå PROBL√àME : Fuite d'encapsulation
class ClasseProblematique:
    def __init__(self):
        self._items = [1, 2, 3]
    
    def get_items(self):
        return self._items  # ‚ùå Retourne la r√©f√©rence directe

obj = ClasseProblematique()
items = obj.get_items()
print(f"Items: {items}")

# Modification de l'ext√©rieur (mauvais !)
items.append(999)
print(f"Items modifi√©s: {obj.get_items()}")
print("‚Üí L'encapsulation est bris√©e !\n")

# ‚úÖ SOLUTION 1 : Retourner une copie
class ClasseCorrecte1:
    def __init__(self):
        self._items = [1, 2, 3]
    
    def get_items(self):
        return self._items.copy()  # ‚úÖ Retourne une copie

obj1 = ClasseCorrecte1()
items1 = obj1.get_items()
items1.append(999)
print(f"ClasseCorrecte1 - Items originaux: {obj1.get_items()}")
print("‚Üí L'original n'est pas affect√©\n")

# ‚úÖ SOLUTION 2 : Retourner un tuple (immuable)
class ClasseCorrecte2:
    def __init__(self):
        self._items = [1, 2, 3]
    
    def get_items(self):
        return tuple(self._items)  # ‚úÖ Retourne un tuple immuable

obj2 = ClasseCorrecte2()
items2 = obj2.get_items()
print(f"ClasseCorrecte2 - Items: {items2}")
try:
    items2.append(999)  # Impossible
except AttributeError:
    print("‚Üí Impossible de modifier un tuple")
````

### Pi√®ge 4 : Confondre encapsulation et masquage

**Erreur** : Penser que mettre `__` partout = bonne encapsulation.

**V√©rit√©** : L'encapsulation, c'est concevoir une **bonne interface publique**, pas juste cacher des choses.

**R√®gle** : Utilisez `_attribut` (prot√©g√©) par d√©faut, `__attribut` (priv√©) rarement.

### Exercice 1 : Compte bancaire s√©curis√© (20 min)

Cr√©ez une classe `CompteBancaireSecurise` avec :

**Attributs priv√©s** :
- `_numero` : num√©ro de compte (lecture seule)
- `_titulaire` : nom du titulaire (lecture/√©criture avec validation)
- `_solde` : solde du compte (lecture seule, modification via m√©thodes)
- `_code_pin` : code PIN √† 4 chiffres (totalement priv√©)

**Propri√©t√©s** :
- `numero` : getter seulement
- `titulaire` : getter + setter (validation : min 2 caract√®res)
- `solde` : getter seulement

**M√©thodes publiques** :
- `verifier_pin(code)` : v√©rifie si le code est correct
- `changer_pin(ancien_code, nouveau_code)` : change le PIN (validation : 4 chiffres)
- `deposer(montant, code)` : d√©pose de l'argent (requiert le PIN)
- `retirer(montant, code)` : retire de l'argent (requiert le PIN)

**Invariants** :
- Le solde ne peut jamais √™tre n√©gatif
- Le PIN doit avoir exactement 4 chiffres
- Toute op√©ration de modification requiert le PIN

````python
# SOLUTION - Exercice 1

class CompteBancaireSecurise:
    """Compte bancaire avec s√©curit√© par code PIN"""
    
    def __init__(self, numero, titulaire, code_pin, solde_initial=0):
        """Initialise un compte s√©curis√©
        
        Args:
            numero (str): Num√©ro de compte
            titulaire (str): Nom du titulaire
            code_pin (str): Code PIN √† 4 chiffres
            solde_initial (float): Solde initial (d√©faut: 0)
        """
        # Validation et initialisation
        self._numero = numero
        self.titulaire = titulaire  # Utilise le setter pour validation
        self._solde = 0
        
        # Validation du PIN
        if not self._valider_pin(code_pin):
            raise ValueError("Le code PIN doit contenir exactement 4 chiffres")
        self.__code_pin = code_pin  # Priv√© avec name mangling
        
        # Validation du solde initial
        if solde_initial < 0:
            raise ValueError("Le solde initial ne peut pas √™tre n√©gatif")
        self._solde = solde_initial
    
    def _valider_pin(self, code):
        """M√©thode priv√©e pour valider le format du PIN"""
        return isinstance(code, str) and len(code) == 4 and code.isdigit()
    
    # PROPRI√âT√âS
    
    @property
    def numero(self):
        """Num√©ro de compte (lecture seule)"""
        return self._numero
    
    @property
    def titulaire(self):
        """Nom du titulaire"""
        return self._titulaire
    
    @titulaire.setter
    def titulaire(self, nom):
        """D√©finit le titulaire avec validation"""
        if not nom or len(nom) < 2:
            raise ValueError("Le nom doit contenir au moins 2 caract√®res")
        self._titulaire = nom.strip()
    
    @property
    def solde(self):
        """Solde du compte (lecture seule)"""
        return self._solde
    
    # M√âTHODES PUBLIQUES
    
    def verifier_pin(self, code):
        """V√©rifie si le code PIN est correct
        
        Args:
            code (str): Code √† v√©rifier
            
        Returns:
            bool: True si correct, False sinon
        """
        return code == self.__code_pin
    
    def changer_pin(self, ancien_code, nouveau_code):
        """Change le code PIN
        
        Args:
            ancien_code (str): Code actuel
            nouveau_code (str): Nouveau code
        """
        if not self.verifier_pin(ancien_code):
            raise ValueError("Code PIN incorrect")
        
        if not self._valider_pin(nouveau_code):
            raise ValueError("Le nouveau PIN doit contenir 4 chiffres")
        
        self.__code_pin = nouveau_code
        print("‚úì Code PIN modifi√© avec succ√®s")
    
    def deposer(self, montant, code):
        """D√©pose de l'argent (requiert le PIN)
        
        Args:
            montant (float): Montant √† d√©poser
            code (str): Code PIN
        """
        if not self.verifier_pin(code):
            print("‚úó Code PIN incorrect. Op√©ration refus√©e.")
            return False
        
        if montant <= 0:
            print("‚úó Le montant doit √™tre positif")
            return False
        
        self._solde += montant
        print(f"‚úì D√©p√¥t de {montant}‚Ç¨. Nouveau solde: {self._solde}‚Ç¨")
        return True
    
    def retirer(self, montant, code):
        """Retire de l'argent (requiert le PIN)
        
        Args:
            montant (float): Montant √† retirer
            code (str): Code PIN
        """
        if not self.verifier_pin(code):
            print("‚úó Code PIN incorrect. Op√©ration refus√©e.")
            return False
        
        if montant <= 0:
            print("‚úó Le montant doit √™tre positif")
            return False
        
        if montant > self._solde:
            print(f"‚úó Solde insuffisant. Disponible: {self._solde}‚Ç¨")
            return False
        
        self._solde -= montant
        print(f"‚úì Retrait de {montant}‚Ç¨. Nouveau solde: {self._solde}‚Ç¨")
        return True
    
    def afficher_info(self):
        """Affiche les informations du compte (sans PIN !)"""
        print(f"\nCompte {self._numero}")
        print(f"Titulaire: {self._titulaire}")
        print(f"Solde: {self._solde}‚Ç¨")
        print(f"PIN: ****")  # Masqu√©

# Test de la classe
print("\n=== Test CompteBancaireSecurise ===")

compte = CompteBancaireSecurise("FR123456", "Alice Dupont", "1234", 1000)
compte.afficher_info()

print("\n--- Op√©rations avec bon code ---")
compte.deposer(500, "1234")
compte.retirer(200, "1234")

print("\n--- Op√©rations avec mauvais code ---")
compte.deposer(100, "9999")
compte.retirer(100, "9999")

print("\n--- Changement de PIN ---")
compte.changer_pin("1234", "5678")
compte.deposer(100, "5678")  # Avec le nouveau code

print("\n--- Tentative d'acc√®s direct au PIN ---")
try:
    print(compte.__code_pin)
except AttributeError:
    print("‚úì Impossible d'acc√©der au PIN directement (bien encapsul√©)")

compte.afficher_info()
````

### Exercice 2 : Identifier les probl√®mes d'encapsulation (10 min)

Le code suivant viole plusieurs principes d'encapsulation. Identifiez les probl√®mes.

````python
# Code avec probl√®mes d'encapsulation

class Produit:
    def __init__(self, nom, prix):
        self.nom = nom
        self.prix = prix  # Public, pas de validation
        self.historique_prix = []  # Liste publique mutable

class Panier:
    def __init__(self):
        self.produits = []  # Liste publique
        self.total = 0  # Public, peut √™tre modifi√© directement
    
    def ajouter(self, produit):
        self.produits.append(produit)
        self.total += produit.prix  # Calcul manuel

# Utilisation (probl√©matique)
p = Produit("Livre", 20)
panier = Panier()
panier.ajouter(p)

# ‚ùå Probl√®mes :
p.prix = -100  # Pas de validation !
panier.total = 999999  # Modification directe !
panier.produits.clear()  # Modification directe mais total non mis √† jour !

print(f"\nProduit: {p.nom} - Prix: {p.prix}‚Ç¨")
print(f"Panier: {len(panier.produits)} produits, Total: {panier.total}‚Ç¨")
print("‚Üí Le panier est dans un √©tat incoh√©rent !")
````

````python
# SOLUTION - Exercice 2 : Code corrig√©

class ProduitCorrige:
    """Produit avec encapsulation"""
    
    def __init__(self, nom, prix):
        self.nom = nom
        self._prix = 0
        self.prix = prix  # Utilise le setter
        self._historique_prix = [prix]  # Priv√©
    
    @property
    def prix(self):
        return self._prix
    
    @prix.setter
    def prix(self, valeur):
        """D√©finit le prix avec validation"""
        if valeur <= 0:
            raise ValueError("Le prix doit √™tre positif")
        self._prix = valeur
        self._historique_prix.append(valeur)
    
    def get_historique_prix(self):
        """Retourne une copie de l'historique"""
        return self._historique_prix.copy()

class PanierCorrige:
    """Panier avec encapsulation"""
    
    def __init__(self):
        self._produits = []  # Priv√©
    
    @property
    def total(self):
        """Total calcul√© √† la vol√©e (pas stock√©)"""
        return sum(p.prix for p in self._produits)
    
    @property
    def nombre_produits(self):
        """Nombre de produits"""
        return len(self._produits)
    
    def ajouter(self, produit):
        """Ajoute un produit"""
        if not isinstance(produit, ProduitCorrige):
            raise TypeError("Doit √™tre un ProduitCorrige")
        self._produits.append(produit)
    
    def retirer(self, produit):
        """Retire un produit"""
        if produit in self._produits:
            self._produits.remove(produit)
    
    def get_produits(self):
        """Retourne une copie de la liste"""
        return self._produits.copy()

# Test de la version corrig√©e
print("\n=== Version corrig√©e ===")

p = ProduitCorrige("Livre", 20)
panier = PanierCorrige()
panier.ajouter(p)

print(f"\nProduit: {p.nom} - Prix: {p.prix}‚Ç¨")
print(f"Panier: {panier.nombre_produits} produit(s), Total: {panier.total}‚Ç¨")

# ‚úÖ Tentative de modification invalide
try:
    p.prix = -100
except ValueError as e:
    print(f"\n‚úì Erreur captur√©e: {e}")

# ‚úÖ Le total est calcul√© automatiquement
p2 = ProduitCorrige("Stylo", 5)
panier.ajouter(p2)
print(f"\nApr√®s ajout: {panier.nombre_produits} produits, Total: {panier.total}‚Ç¨")

# ‚úÖ Modification de la copie n'affecte pas l'original
produits_copie = panier.get_produits()
produits_copie.clear()
print(f"\nApr√®s clear de la copie: Panier toujours {panier.nombre_produits} produits")

print("\n‚úì L'encapsulation garantit la coh√©rence !")
````

### Exercice 3 : Concevoir une interface publique (15 min)

Concevez une classe `GestionnaireStock` pour g√©rer un stock de produits.

**Contraintes** :
- Le stock ne peut jamais √™tre n√©gatif
- Chaque modification du stock doit √™tre trac√©e
- On doit pouvoir consulter l'historique
- La quantit√© initiale doit √™tre valid√©e

**Interface publique √† concevoir** :
- Quels attributs doivent √™tre publics ?
- Quelles m√©thodes exposer ?
- Comment garantir les invariants ?

````python
# SOLUTION - Exercice 3

from datetime import datetime

class GestionnaireStock:
    """G√®re le stock d'un produit avec tra√ßabilit√©
    
    INVARIANT: La quantit√© en stock est toujours >= 0
    """
    
    def __init__(self, nom_produit, quantite_initiale=0):
        """Initialise le gestionnaire de stock
        
        Args:
            nom_produit (str): Nom du produit
            quantite_initiale (int): Quantit√© initiale (d√©faut: 0)
        """
        if quantite_initiale < 0:
            raise ValueError("La quantit√© initiale ne peut pas √™tre n√©gative")
        
        self._nom_produit = nom_produit  # Lecture seule
        self._quantite = quantite_initiale
        self._historique = []  # Tra√ßabilit√©
        self._ajouter_historique("Initialisation", quantite_initiale, 0)
    
    def _ajouter_historique(self, operation, quantite, ancienne_quantite):
        """M√©thode priv√©e pour ajouter une entr√©e dans l'historique"""
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        self._historique.append({
            'timestamp': timestamp,
            'operation': operation,
            'quantite': quantite,
            'stock_avant': ancienne_quantite,
            'stock_apres': self._quantite
        })
    
    # PROPRI√âT√âS
    
    @property
    def nom_produit(self):
        """Nom du produit (lecture seule)"""
        return self._nom_produit
    
    @property
    def quantite(self):
        """Quantit√© en stock (lecture seule)"""
        return self._quantite
    
    @property
    def est_disponible(self):
        """V√©rifie si le produit est en stock"""
        return self._quantite > 0
    
    # M√âTHODES PUBLIQUES
    
    def approvisionner(self, quantite):
        """Ajoute du stock
        
        Args:
            quantite (int): Quantit√© √† ajouter
        """
        if quantite <= 0:
            raise ValueError("La quantit√© doit √™tre positive")
        
        ancienne_quantite = self._quantite
        self._quantite += quantite
        self._ajouter_historique("Approvisionnement", quantite, ancienne_quantite)
        print(f"‚úì +{quantite} {self._nom_produit}. Stock: {self._quantite}")
    
    def vendre(self, quantite):
        """Retire du stock (vente)
        
        Args:
            quantite (int): Quantit√© √† vendre
            
        Returns:
            bool: True si vente r√©ussie, False sinon
        """
        if quantite <= 0:
            print("‚úó La quantit√© doit √™tre positive")
            return False
        
        if quantite > self._quantite:
            print(f"‚úó Stock insuffisant. Disponible: {self._quantite}")
            return False
        
        ancienne_quantite = self._quantite
        self._quantite -= quantite
        self._ajouter_historique("Vente", quantite, ancienne_quantite)
        print(f"‚úì Vente de {quantite} {self._nom_produit}. Stock restant: {self._quantite}")
        
        # Alerte si stock faible
        if self._quantite < 10:
            print(f"  ‚ö†Ô∏è  Alerte: Stock faible ({self._quantite})")
        
        return True
    
    def ajuster_stock(self, nouvelle_quantite, raison):
        """Ajuste le stock (inventaire, casse, etc.)
        
        Args:
            nouvelle_quantite (int): Nouvelle quantit√©
            raison (str): Raison de l'ajustement
        """
        if nouvelle_quantite < 0:
            raise ValueError("La quantit√© ne peut pas √™tre n√©gative")
        
        ancienne_quantite = self._quantite
        diff = nouvelle_quantite - ancienne_quantite
        self._quantite = nouvelle_quantite
        
        operation = f"Ajustement ({raison})"
        self._ajouter_historique(operation, diff, ancienne_quantite)
        print(f"‚úì Ajustement: {ancienne_quantite} ‚Üí {nouvelle_quantite} ({raison})")
    
    def get_historique(self):
        """Retourne une copie de l'historique"""
        return self._historique.copy()
    
    def afficher_historique(self, limite=10):
        """Affiche l'historique des op√©rations
        
        Args:
            limite (int): Nombre d'entr√©es √† afficher (d√©faut: 10)
        """
        print(f"\n{'='*70}")
        print(f"Historique de {self._nom_produit}")
        print(f"{'='*70}")
        
        for entry in self._historique[-limite:]:
            print(f"[{entry['timestamp']}] {entry['operation']}")
            print(f"  Quantit√©: {entry['quantite']:+d} | "
                  f"Stock: {entry['stock_avant']} ‚Üí {entry['stock_apres']}")
        
        print(f"{'='*70}\n")
    
    def __repr__(self):
        return f"GestionnaireStock('{self._nom_produit}', {self._quantite})"

# Test de la classe
print("\n=== Test GestionnaireStock ===")

stock = GestionnaireStock("Ordinateur portable", 50)
print(f"\nStock initial: {stock}")
print(f"Disponible ? {stock.est_disponible}")

# Op√©rations
print("\n--- Op√©rations ---")
stock.vendre(5)
stock.vendre(10)
stock.approvisionner(20)
stock.vendre(30)
stock.ajuster_stock(15, "Inventaire annuel")
stock.vendre(8)

# Affichage de l'historique
stock.afficher_historique()

print(f"Stock final: {stock}")
````
