## 2. Les Attributs : L'Ã‰tat d'un Objet

### 2.1 Qu'est-ce qu'un attribut ?

Un **attribut** (aussi appelÃ© **propriÃ©tÃ©**, **champ** ou **variable d'instance**) est une **donnÃ©e** qui appartient Ã  un objet.

Les attributs reprÃ©sentent l'**Ã‰TAT** de l'objet Ã  un instant donnÃ©.

### 2.2 Types d'attributs

Il existe deux grandes catÃ©gories d'attributs :

#### Attributs d'instance (instance attributes)
- **Propres Ã  chaque objet**
- Chaque instance a sa propre copie
- Peuvent avoir des valeurs diffÃ©rentes pour chaque objet
- DÃ©finis gÃ©nÃ©ralement dans le constructeur (`__init__`)

**Exemple** : Chaque voiture a sa propre couleur, son propre kilomÃ©trage

#### Attributs de classe (class attributes)
- **PartagÃ©s par toutes les instances** de la classe
- Une seule copie pour toute la classe
- MÃªme valeur pour tous les objets (sauf si explicitement modifiÃ©)
- DÃ©finis directement dans la classe (hors des mÃ©thodes)

**Exemple** : Toutes les voitures ont 4 roues (constante de la classe)

### 2.3 VisibilitÃ© conceptuelle

En thÃ©orie (nous verrons l'implÃ©mentation Python dans le notebook suivant), les attributs peuvent avoir diffÃ©rents niveaux de visibilitÃ© :

- **Public** : Accessible de partout
- **PrivÃ©** : Accessible uniquement Ã  l'intÃ©rieur de la classe
- **ProtÃ©gÃ©** : Accessible dans la classe et ses sous-classes

### 2.4 Bonnes pratiques

1. **Nommer clairement** : `age`, `nom_complet`, `solde_compte`
2. **Ã‰viter les attributs publics mutables** si possible
3. **Initialiser dans le constructeur** pour garantir la cohÃ©rence
4. **Documenter** les attributs dans la docstring de la classe

## 3. Les MÃ©thodes : Le Comportement d'un Objet

### 3.1 Qu'est-ce qu'une mÃ©thode ?

Une **mÃ©thode** est une **fonction** qui appartient Ã  une classe. Elle dÃ©finit les **actions** qu'un objet peut effectuer.

Les mÃ©thodes reprÃ©sentent le **COMPORTEMENT** de l'objet.

### 3.2 Types de mÃ©thodes

#### MÃ©thodes d'instance (instance methods)
- **OpÃ¨rent sur un objet spÃ©cifique**
- Ont accÃ¨s Ã  l'instance via `self` (en Python)
- Peuvent lire et modifier l'Ã©tat de l'objet
- Les plus courantes

**Exemple** : `voiture.demarrer()`, `compte.deposer(100)`

#### MÃ©thodes de classe (class methods)
- **OpÃ¨rent sur la classe elle-mÃªme**, pas sur une instance
- Ont accÃ¨s Ã  la classe via `cls` (en Python)
- Utiles pour les factory methods (mÃ©thodes de crÃ©ation)
- MarquÃ©es avec `@classmethod` en Python

**Exemple** : `Date.from_string("2024-01-01")` pour crÃ©er une date Ã  partir d'une chaÃ®ne

#### MÃ©thodes statiques (static methods)
- **N'opÃ¨rent ni sur l'instance ni sur la classe**
- Fonctions utilitaires liÃ©es conceptuellement Ã  la classe
- Pas d'accÃ¨s Ã  `self` ou `cls`
- MarquÃ©es avec `@staticmethod` en Python

**Exemple** : `Math.est_pair(nombre)` pour vÃ©rifier si un nombre est pair

### 3.3 Le constructeur : une mÃ©thode spÃ©ciale

Le **constructeur** est une mÃ©thode spÃ©ciale appelÃ©e lors de la **crÃ©ation** d'un objet.

**RÃ´les** :
1. **Initialiser** les attributs d'instance
2. **Valider** les paramÃ¨tres de crÃ©ation
3. **Configurer** l'Ã©tat initial de l'objet
4. **Effectuer** des opÃ©rations nÃ©cessaires au dÃ©marrage

**En Python** : `__init__(self, ...)`

### 3.4 MÃ©thodes vs Fonctions

| Fonction | MÃ©thode |
|----------|----------|
| IndÃ©pendante | LiÃ©e Ã  une classe |
| `fonction(objet, params)` | `objet.methode(params)` |
| Pas de `self` | AccÃ¨s Ã  `self` |
| ProcÃ©durale | OrientÃ©e objet |

### 3.5 Bonnes pratiques

1. **Une mÃ©thode = une responsabilitÃ©** (Single Responsibility Principle)
2. **Noms verbaux** : `calculer_total()`, `valider_email()`
3. **Documenter** avec des docstrings

````python
# Exemple 1 : Classe simple avec attributs d'instance

class Livre:
    """ReprÃ©sente un livre avec ses informations de base"""
    
    def __init__(self, titre, auteur, annee):
        """Constructeur : initialise un nouveau livre
        
        Args:
            titre (str): Le titre du livre
            auteur (str): Le nom de l'auteur
            annee (int): L'annÃ©e de publication
        """
        # Attributs d'instance
        self.titre = titre
        self.auteur = auteur
        self.annee = annee
        self.est_emprunte = False  # Ã‰tat initial par dÃ©faut
    
    def emprunter(self):
        """Marque le livre comme empruntÃ©"""
        if not self.est_emprunte:
            self.est_emprunte = True
            print(f"'{self.titre}' a Ã©tÃ© empruntÃ©")
        else:
            print(f"'{self.titre}' est dÃ©jÃ  empruntÃ©")
    
    def retourner(self):
        """Marque le livre comme retournÃ©"""
        if self.est_emprunte:
            self.est_emprunte = False
            print(f"'{self.titre}' a Ã©tÃ© retournÃ©")
        else:
            print(f"'{self.titre}' n'Ã©tait pas empruntÃ©")
    
    def afficher_info(self):
        """Affiche les informations du livre"""
        statut = "EmpruntÃ©" if self.est_emprunte else "Disponible"
        print(f"'{self.titre}' de {self.auteur} ({self.annee}) - {statut}")

# Utilisation
print("=== Exemple 1 : Attributs d'instance ===")

livre1 = Livre("Le Petit Prince", "Antoine de Saint-ExupÃ©ry", 1943)
livre2 = Livre("1984", "George Orwell", 1949)

livre1.afficher_info()
livre2.afficher_info()

print()
livre1.emprunter()
livre1.afficher_info()
livre2.afficher_info()  # Livre2 n'est pas affectÃ©

print()
livre1.retourner()
livre1.afficher_info()
````

````python
# Exemple 2 : Attributs de classe vs attributs d'instance

class Employe:
    """ReprÃ©sente un employÃ© d'une entreprise"""
    
    # ATTRIBUT DE CLASSE : partagÃ© par toutes les instances
    entreprise = "TechCorp"  # Tous les employÃ©s travaillent pour la mÃªme entreprise
    nombre_employes = 0       # Compteur du nombre total d'employÃ©s
    
    def __init__(self, nom, poste, salaire):
        """Initialise un employÃ©"""
        # ATTRIBUTS D'INSTANCE : propres Ã  chaque employÃ©
        self.nom = nom
        self.poste = poste
        self.salaire = salaire
        
        # IncrÃ©menter le compteur de classe
        Employe.nombre_employes += 1
    
    def afficher_info(self):
        """Affiche les informations de l'employÃ©"""
        print(f"{self.nom} - {self.poste} chez {Employe.entreprise} - {self.salaire}â‚¬")
    
    def augmenter_salaire(self, pourcentage):
        """Augmente le salaire de l'employÃ©"""
        self.salaire = self.salaire * (1 + pourcentage / 100)
        print(f"Salaire de {self.nom} augmentÃ© de {pourcentage}%")
    
    @classmethod
    def changer_entreprise(cls, nouveau_nom):
        """Change le nom de l'entreprise pour tous les employÃ©s"""
        cls.entreprise = nouveau_nom
        print(f"L'entreprise s'appelle maintenant : {nouveau_nom}")
    
    @classmethod
    def afficher_statistiques(cls):
        """Affiche les statistiques globales"""
        print(f"Nombre total d'employÃ©s : {cls.nombre_employes}")
        print(f"Entreprise : {cls.entreprise}")

# Utilisation
print("\n=== Exemple 2 : Attributs de classe ===")

emp1 = Employe("Alice Dupont", "DÃ©veloppeuse", 45000)
emp2 = Employe("Bob Martin", "Designer", 40000)
emp3 = Employe("Charlie Durand", "Chef de projet", 55000)

print("\nInformations individuelles :")
emp1.afficher_info()
emp2.afficher_info()
emp3.afficher_info()

print("\nStatistiques globales :")
Employe.afficher_statistiques()

print("\nChangement du nom de l'entreprise :")
Employe.changer_entreprise("MegaCorp")

print("\nTous les employÃ©s sont affectÃ©s :")
emp1.afficher_info()
emp2.afficher_info()
emp3.afficher_info()

print("\nMais les salaires sont individuels :")
emp1.augmenter_salaire(10)
emp1.afficher_info()
emp2.afficher_info()  # Salaire de Bob inchangÃ©
````

````python
# Exemple 3 : MÃ©thodes d'instance, de classe et statiques

from datetime import datetime

class Personne:
    """ReprÃ©sente une personne avec diffÃ©rents types de mÃ©thodes"""
    
    def __init__(self, nom, prenom, annee_naissance):
        """Constructeur"""
        self.nom = nom
        self.prenom = prenom
        self.annee_naissance = annee_naissance
    
    # MÃ‰THODE D'INSTANCE : opÃ¨re sur l'objet spÃ©cifique
    def calculer_age(self):
        """Calcule l'Ã¢ge de la personne"""
        annee_actuelle = datetime.now().year
        return annee_actuelle - self.annee_naissance
    
    def afficher(self):
        """Affiche les informations de la personne"""
        print(f"{self.prenom} {self.nom}, {self.calculer_age()} ans")
    
    # MÃ‰THODE DE CLASSE : factory method pour crÃ©er des objets
    @classmethod
    def from_string(cls, chaine):
        """CrÃ©e une personne Ã  partir d'une chaÃ®ne 'Nom,Prenom,Annee'
        
        Exemple: Personne.from_string("Dupont,Alice,1990")
        """
        nom, prenom, annee = chaine.split(',')
        return cls(nom, prenom, int(annee))
    
    @classmethod
    def from_dict(cls, data):
        """CrÃ©e une personne Ã  partir d'un dictionnaire"""
        return cls(data['nom'], data['prenom'], data['annee_naissance'])
    
    # MÃ‰THODE STATIQUE : fonction utilitaire liÃ©e conceptuellement Ã  la classe
    @staticmethod
    def est_majeur(age):
        """VÃ©rifie si un Ã¢ge correspond Ã  une personne majeure"""
        return age >= 18
    
    @staticmethod
    def formater_nom_complet(nom, prenom):
        """Formate un nom complet en majuscules"""
        return f"{prenom.upper()} {nom.upper()}"

# Utilisation
print("\n=== Exemple 3 : Types de mÃ©thodes ===")

# CrÃ©ation classique (via constructeur)
p1 = Personne("Dupont", "Alice", 1990)
print("\nCrÃ©ation classique :")
p1.afficher()

# CrÃ©ation via mÃ©thode de classe (factory method)
p2 = Personne.from_string("Martin,Bob,2005")
print("\nCrÃ©ation via from_string :")
p2.afficher()

p3 = Personne.from_dict({
    'nom': 'Durand',
    'prenom': 'Charlie',
    'annee_naissance': 2010
})
print("\nCrÃ©ation via from_dict :")
p3.afficher()

# Utilisation de mÃ©thodes statiques
print("\nMÃ©thodes statiques :")
print(f"Alice est majeure : {Personne.est_majeur(p1.calculer_age())}")
print(f"Bob est majeur : {Personne.est_majeur(p2.calculer_age())}")
print(f"Charlie est majeur : {Personne.est_majeur(p3.calculer_age())}")

print(f"\nNom formatÃ© : {Personne.formater_nom_complet('dupont', 'alice')}")
````

````python
# Exemple 4 : Le constructeur et la validation

class CompteBancaire:
    """Compte bancaire avec validation dans le constructeur"""
    
    SOLDE_MINIMUM = 0  # Attribut de classe
    
    def __init__(self, numero, titulaire, solde_initial=0):
        """Initialise un compte avec validation
        
        Args:
            numero (str): NumÃ©ro du compte
            titulaire (str): Nom du titulaire
            solde_initial (float): Solde de dÃ©part (dÃ©faut: 0)
            
        Raises:
            ValueError: Si les paramÃ¨tres sont invalides
        """
        # VALIDATION dans le constructeur
        if not numero or not isinstance(numero, str):
            raise ValueError("Le numÃ©ro de compte doit Ãªtre une chaÃ®ne non vide")
        
        if not titulaire or not isinstance(titulaire, str):
            raise ValueError("Le titulaire doit Ãªtre une chaÃ®ne non vide")
        
        if solde_initial < self.SOLDE_MINIMUM:
            raise ValueError(f"Le solde initial ne peut pas Ãªtre nÃ©gatif")
        
        # Initialisation des attributs (seulement si validation OK)
        self.numero = numero
        self.titulaire = titulaire
        self.solde = solde_initial
        self.historique = []  # Liste pour garder l'historique des opÃ©rations
        
        # Enregistrer la crÃ©ation dans l'historique
        self._ajouter_historique(f"Compte crÃ©Ã© avec solde initial: {solde_initial}â‚¬")
    
    def _ajouter_historique(self, message):
        """MÃ©thode privÃ©e pour ajouter une entrÃ©e dans l'historique"""
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        self.historique.append(f"[{timestamp}] {message}")
    
    def deposer(self, montant):
        """DÃ©pose de l'argent"""
        if montant <= 0:
            print("Le montant doit Ãªtre positif")
            return
        
        self.solde += montant
        self._ajouter_historique(f"DÃ©pÃ´t: +{montant}â‚¬")
        print(f"DÃ©pÃ´t de {montant}â‚¬ effectuÃ©. Nouveau solde: {self.solde}â‚¬")
    
    def retirer(self, montant):
        """Retire de l'argent"""
        if montant <= 0:
            print("Le montant doit Ãªtre positif")
            return
        
        if self.solde - montant < self.SOLDE_MINIMUM:
            print(f"Solde insuffisant. Solde actuel: {self.solde}â‚¬")
            return
        
        self.solde -= montant
        self._ajouter_historique(f"Retrait: -{montant}â‚¬")
        print(f"Retrait de {montant}â‚¬ effectuÃ©. Nouveau solde: {self.solde}â‚¬")
    
    def afficher_historique(self):
        """Affiche l'historique des opÃ©rations"""
        print(f"\nHistorique du compte {self.numero} ({self.titulaire}):")
        for operation in self.historique:
            print(f"  {operation}")
        print(f"Solde actuel: {self.solde}â‚¬")

# Utilisation
print("\n=== Exemple 4 : Constructeur et validation ===")

try:
    # CrÃ©ation rÃ©ussie
    compte = CompteBancaire("FR123456", "Alice Dupont", 1000)
    print("âœ“ Compte crÃ©Ã© avec succÃ¨s")
    
    compte.deposer(500)
    compte.retirer(200)
    compte.deposer(100)
    
    compte.afficher_historique()
    
except ValueError as e:
    print(f"âœ— Erreur: {e}")

print("\nTentative de crÃ©ation avec solde nÃ©gatif :")
try:
    compte_invalide = CompteBancaire("FR999", "Bob", -100)
except ValueError as e:
    print(f"âœ— Erreur capturÃ©e: {e}")
````

### PiÃ¨ge 4 : Confusion entre mÃ©thodes de classe et mÃ©thodes statiques

**Question** : Quand utiliser `@classmethod` vs `@staticmethod` ?

**RÃ¨gle** :
- `@classmethod` : Quand vous avez besoin d'accÃ©der Ã  la classe (`cls`)
- `@staticmethod` : Quand c'est juste une fonction utilitaire liÃ©e conceptuellement Ã  la classe

````python
# DÃ©monstration du piÃ¨ge 4

class MathUtils:
    """Classe utilitaire pour opÃ©rations mathÃ©matiques"""
    
    PI = 3.14159
    
    @staticmethod
    def est_pair(n):
        """MÃ©thode statique : pas besoin de la classe"""
        return n % 2 == 0
    
    @classmethod
    def aire_cercle(cls, rayon):
        """MÃ©thode de classe : utilise un attribut de classe (PI)"""
        return cls.PI * rayon ** 2

print("\nUtilisation des mÃ©thodes :")
print(f"5 est pair ? {MathUtils.est_pair(5)}")
print(f"Aire d'un cercle de rayon 3 : {MathUtils.aire_cercle(3):.2f}")

print("\nRemarque :")
print("- est_pair() pourrait Ãªtre une fonction indÃ©pendante")
print("- aire_cercle() a besoin de PI (attribut de classe)")
````

### Exercice 1 : ModÃ©liser un objet du quotidien (20 min)

**Objectif** : CrÃ©er une classe `Smartphone` avec attributs et mÃ©thodes appropriÃ©s.

**SpÃ©cifications** :
- **Attributs d'instance** : marque, modÃ¨le, capacitÃ©_batterie (en %), stockage_utilise (en Go), stockage_total (en Go)
- **Attribut de classe** : taille_ecran (tous les smartphones modernes ont des Ã©crans de 6 pouces)
- **MÃ©thodes** :
  - `utiliser(duree)` : rÃ©duit la batterie de 10% par heure d'utilisation
  - `recharger(duree)` : augmente la batterie de 20% par heure de charge (max 100%)
  - `installer_app(taille)` : installe une app et rÃ©duit le stockage disponible
  - `afficher_etat()` : affiche les informations du smartphone
- **Validation** : La batterie ne peut pas Ãªtre nÃ©gative, le stockage utilisÃ© ne peut pas dÃ©passer le total

**Bonus** : Ajoutez une mÃ©thode de classe pour crÃ©er un smartphone Ã  partir d'un dictionnaire.

````python
# SOLUTION - Exercice 1 : Classe Smartphone

class Smartphone:
    """ReprÃ©sente un smartphone avec ses caractÃ©ristiques et fonctionnalitÃ©s"""
    
    # Attribut de classe
    taille_ecran = 6.0  # pouces
    
    def __init__(self, marque, modele, stockage_total, batterie_initiale=100):
        """Initialise un smartphone
        
        Args:
            marque (str): Marque du smartphone
            modele (str): ModÃ¨le du smartphone
            stockage_total (int): Stockage total en Go
            batterie_initiale (int): Niveau de batterie initial (dÃ©faut: 100%)
        """
        self.marque = marque
        self.modele = modele
        self.capacite_batterie = min(100, max(0, batterie_initiale))  # Entre 0 et 100
        self.stockage_utilise = 0
        self.stockage_total = stockage_total
    
    def utiliser(self, duree):
        """Utilise le smartphone pendant une certaine durÃ©e
        
        Args:
            duree (float): DurÃ©e d'utilisation en heures
        """
        perte = duree * 10
        self.capacite_batterie = max(0, self.capacite_batterie - perte)
        
        if self.capacite_batterie == 0:
            print("âš ï¸  Batterie Ã©puisÃ©e ! Veuillez recharger.")
        else:
            print(f"Utilisation pendant {duree}h. Batterie restante: {self.capacite_batterie}%")
    
    def recharger(self, duree):
        """Recharge le smartphone pendant une certaine durÃ©e
        
        Args:
            duree (float): DurÃ©e de charge en heures
        """
        gain = duree * 20
        self.capacite_batterie = min(100, self.capacite_batterie + gain)
        print(f"Charge pendant {duree}h. Batterie: {self.capacite_batterie}%")
    
    def installer_app(self, nom_app, taille):
        """Installe une application
        
        Args:
            nom_app (str): Nom de l'application
            taille (float): Taille de l'app en Go
        
        Returns:
            bool: True si installation rÃ©ussie, False sinon
        """
        stockage_disponible = self.stockage_total - self.stockage_utilise
        
        if taille > stockage_disponible:
            print(f"âŒ Stockage insuffisant pour {nom_app} ({taille}Go)")
            print(f"   Disponible: {stockage_disponible}Go")
            return False
        
        self.stockage_utilise += taille
        print(f"âœ“ {nom_app} installÃ©e ({taille}Go)")
        return True
    
    def afficher_etat(self):
        """Affiche l'Ã©tat actuel du smartphone"""
        stockage_disponible = self.stockage_total - self.stockage_utilise
        print(f"\n{'='*50}")
        print(f"  {self.marque} {self.modele}")
        print(f"{'='*50}")
        print(f"  Ã‰cran: {self.taille_ecran} pouces")
        print(f"  Batterie: {self.capacite_batterie}%")
        print(f"  Stockage: {self.stockage_utilise}/{self.stockage_total}Go utilisÃ©s")
        print(f"  Disponible: {stockage_disponible}Go")
        print(f"{'='*50}\n")
    
    @classmethod
    def from_dict(cls, data):
        """CrÃ©e un smartphone Ã  partir d'un dictionnaire
        
        Args:
            data (dict): Dictionnaire avec les clÃ©s 'marque', 'modele', 'stockage_total'
        """
        return cls(
            data['marque'],
            data['modele'],
            data['stockage_total'],
            data.get('batterie_initiale', 100)
        )

# Test de la classe
print("\n=== Test de la classe Smartphone ===")

# CrÃ©ation
phone = Smartphone("Apple", "iPhone 15", 128)
phone.afficher_etat()

# Installation d'apps
phone.installer_app("Instagram", 0.5)
phone.installer_app("WhatsApp", 0.3)
phone.installer_app("Spotify", 0.4)

# Utilisation
print()
phone.utiliser(3)  # 3 heures d'utilisation
phone.utiliser(5)  # 5 heures supplÃ©mentaires

# Recharge
print()
phone.recharger(2)  # 2 heures de charge

phone.afficher_etat()

# CrÃ©ation via dictionnaire
phone2 = Smartphone.from_dict({
    'marque': 'Samsung',
    'modele': 'Galaxy S24',
    'stockage_total': 256,
    'batterie_initiale': 75
})
phone2.afficher_etat()
````
