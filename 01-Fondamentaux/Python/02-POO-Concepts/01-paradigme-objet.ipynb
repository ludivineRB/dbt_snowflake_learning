## 8. Exemples en Python : Comparaison ProcÃ©dural vs Objet

Voyons concrÃ¨tement la diffÃ©rence entre une approche procÃ©durale et une approche orientÃ©e objet.

````python
# Approche PROCÃ‰DURALE : Gestion d'un compte bancaire

# Les donnÃ©es sont sÃ©parÃ©es des opÃ©rations
compte1_numero = "123456"
compte1_titulaire = "Alice Dupont"
compte1_solde = 1000.0

compte2_numero = "789012"
compte2_titulaire = "Bob Martin"
compte2_solde = 500.0

# Fonctions qui manipulent les donnÃ©es
def deposer(solde, montant):
    """DÃ©pose de l'argent sur un compte"""
    if montant > 0:
        return solde + montant
    else:
        print("Montant invalide")
        return solde

def retirer(solde, montant):
    """Retire de l'argent d'un compte"""
    if montant > 0 and solde >= montant:
        return solde - montant
    else:
        print("Retrait impossible")
        return solde

def afficher_compte(numero, titulaire, solde):
    """Affiche les informations d'un compte"""
    print(f"Compte {numero} - {titulaire} : {solde}â‚¬")

# Utilisation
print("=== Approche ProcÃ©durale ===")
afficher_compte(compte1_numero, compte1_titulaire, compte1_solde)
compte1_solde = deposer(compte1_solde, 500)
compte1_solde = retirer(compte1_solde, 200)
afficher_compte(compte1_numero, compte1_titulaire, compte1_solde)

# ProblÃ¨mes de cette approche :
# 1. Multiplication des variables (compte1_x, compte2_x...)
# 2. Aucun lien explicite entre donnÃ©es et opÃ©rations
# 3. Risque d'incohÃ©rence (oublier de mettre Ã  jour une variable)
# 4. Pas de protection des donnÃ©es (on peut modifier solde directement)
````

````python
# Approche ORIENTÃ‰E OBJET : Gestion d'un compte bancaire

class CompteBancaire:
    """ReprÃ©sente un compte bancaire avec ses opÃ©rations"""
    
    def __init__(self, numero, titulaire, solde_initial=0.0):
        """Initialise un nouveau compte"""
        self.numero = numero        # Ã‰TAT
        self.titulaire = titulaire  # Ã‰TAT
        self.solde = solde_initial  # Ã‰TAT
    
    def deposer(self, montant):     # COMPORTEMENT
        """DÃ©pose de l'argent sur le compte"""
        if montant > 0:
            self.solde += montant
            print(f"DÃ©pÃ´t de {montant}â‚¬ effectuÃ©")
        else:
            print("Montant invalide")
    
    def retirer(self, montant):     # COMPORTEMENT
        """Retire de l'argent du compte"""
        if montant > 0 and self.solde >= montant:
            self.solde -= montant
            print(f"Retrait de {montant}â‚¬ effectuÃ©")
        else:
            print("Retrait impossible")
    
    def afficher(self):             # COMPORTEMENT
        """Affiche les informations du compte"""
        print(f"Compte {self.numero} - {self.titulaire} : {self.solde}â‚¬")

# Utilisation
print("\n=== Approche OrientÃ©e Objet ===")

# CrÃ©ation d'objets (instances)
compte1 = CompteBancaire("123456", "Alice Dupont", 1000.0)
compte2 = CompteBancaire("789012", "Bob Martin", 500.0)

# Envoi de messages aux objets
compte1.afficher()
compte1.deposer(500)   # Message : "dÃ©pose 500"
compte1.retirer(200)   # Message : "retire 200"
compte1.afficher()

print()
compte2.afficher()

# Avantages de cette approche :
# 1. DonnÃ©es et opÃ©rations regroupÃ©es dans une seule entitÃ©
# 2. Code plus lisible et proche du langage mÃ©tier
# 3. RÃ©utilisabilitÃ© : on peut crÃ©er autant de comptes que nÃ©cessaire
# 4. Chaque objet a son propre Ã©tat (IDENTITÃ‰)
````

````python
# DÃ©monstration des 3 piliers : Ã‰TAT, COMPORTEMENT, IDENTITÃ‰

print("=== Les 3 Piliers d'un Objet ===")

# CrÃ©ons deux comptes avec le mÃªme Ã©tat initial
compte_alice = CompteBancaire("111111", "Alice", 1000.0)
compte_bob = CompteBancaire("222222", "Bob", 1000.0)

print("\n1. Ã‰TAT - Chaque objet a son propre Ã©tat :")
print(f"Compte Alice - Solde: {compte_alice.solde}â‚¬")
print(f"Compte Bob - Solde: {compte_bob.solde}â‚¬")

print("\n2. COMPORTEMENT - Les objets peuvent effectuer des actions :")
compte_alice.deposer(500)
print(f"Compte Alice aprÃ¨s dÃ©pÃ´t - Solde: {compte_alice.solde}â‚¬")
print(f"Compte Bob (non modifiÃ©) - Solde: {compte_bob.solde}â‚¬")

print("\n3. IDENTITÃ‰ - Chaque objet est unique :")
print(f"ID de l'objet compte_alice : {id(compte_alice)}")
print(f"ID de l'objet compte_bob : {id(compte_bob)}")
print(f"Les objets sont diffÃ©rents : {compte_alice is not compte_bob}")

# MÃªme si on met le mÃªme Ã©tat, ce sont toujours des objets distincts
compte_bob.deposer(500)  # Bob a maintenant le mÃªme solde qu'Alice
print(f"\nMÃªme solde ({compte_alice.solde}â‚¬ == {compte_bob.solde}â‚¬) mais objets diffÃ©rents !")
print(f"Ã‰galitÃ© d'Ã©tat : {compte_alice.solde == compte_bob.solde}")
print(f"Ã‰galitÃ© d'identitÃ© : {compte_alice is compte_bob}")
````

````python
# DÃ©monstration de l'ABSTRACTION

print("=== Principe d'Abstraction ===")

# En utilisant CompteBancaire, on n'a pas besoin de savoir :
# - Comment le solde est stockÃ© en mÃ©moire
# - Comment les vÃ©rifications sont effectuÃ©es
# - Comment les logs sont gÃ©rÃ©s
# On utilise simplement l'INTERFACE publique (les mÃ©thodes)

compte = CompteBancaire("999999", "Charlie", 100.0)

# Interface simple pour l'utilisateur
compte.deposer(50)   # Abstraction : on ne sait pas comment c'est fait
compte.retirer(30)   # On envoie juste un message Ã  l'objet

# Comparaison avec l'approche procÃ©durale :
# solde = deposer(solde, 50)
# solde = retirer(solde, 30)
# Il faut gÃ©rer manuellement le passage de donnÃ©es

print("\nAvec la POO, l'objet gÃ¨re lui-mÃªme son Ã©tat.")
print("C'est le principe d'ENCAPSULATION (que nous verrons plus tard).")
````

### Exercice 3 : ProcÃ©dural vs Objet (20 min)

**Contexte** : Vous devez gÃ©rer une liste de rectangles et calculer leur aire.

**Partie A** : ImplÃ©mentez une solution **procÃ©durale**
- Utilisez des listes ou dictionnaires pour stocker les donnÃ©es
- CrÃ©ez des fonctions pour calculer l'aire et afficher les rectangles

**Partie B** : ImplÃ©mentez une solution **orientÃ©e objet**
- CrÃ©ez une classe `Rectangle`
- Ajoutez des mÃ©thodes appropriÃ©es

**Partie C** : Comparez les deux approches
- Quels sont les avantages de chaque approche ?
- Laquelle prÃ©fÃ©rez-vous et pourquoi ?

````python
# SOLUTION - Exercice 3 Partie A : Approche PROCÃ‰DURALE

print("=== APPROCHE PROCÃ‰DURALE ===")

# DonnÃ©es : listes parallÃ¨les (ou dictionnaires)
rectangles_largeurs = [10, 5, 8]
rectangles_hauteurs = [20, 10, 12]
rectangles_noms = ["Rect1", "Rect2", "Rect3"]

def calculer_aire(largeur, hauteur):
    """Calcule l'aire d'un rectangle"""
    return largeur * hauteur

def calculer_perimetre(largeur, hauteur):
    """Calcule le pÃ©rimÃ¨tre d'un rectangle"""
    return 2 * (largeur + hauteur)

def afficher_rectangle(nom, largeur, hauteur):
    """Affiche les informations d'un rectangle"""
    aire = calculer_aire(largeur, hauteur)
    perimetre = calculer_perimetre(largeur, hauteur)
    print(f"{nom}: {largeur}x{hauteur} - Aire: {aire}, PÃ©rimÃ¨tre: {perimetre}")

def afficher_tous_rectangles(noms, largeurs, hauteurs):
    """Affiche tous les rectangles"""
    for nom, largeur, hauteur in zip(noms, largeurs, hauteurs):
        afficher_rectangle(nom, largeur, hauteur)

# Utilisation
afficher_tous_rectangles(rectangles_noms, rectangles_largeurs, rectangles_hauteurs)

# InconvÃ©nients visibles :
# - 3 listes Ã  maintenir synchronisÃ©es
# - Facile de faire des erreurs (mauvais index)
# - Passage de multiples paramÃ¨tres aux fonctions
````

````python
# SOLUTION - Exercice 3 Partie B : Approche ORIENTÃ‰E OBJET

print("\n=== APPROCHE ORIENTÃ‰E OBJET ===")

class Rectangle:
    """ReprÃ©sente un rectangle avec ses dimensions et opÃ©rations"""
    
    def __init__(self, nom, largeur, hauteur):
        """Initialise un rectangle"""
        self.nom = nom           # Ã‰TAT
        self.largeur = largeur   # Ã‰TAT
        self.hauteur = hauteur   # Ã‰TAT
    
    def calculer_aire(self):     # COMPORTEMENT
        """Calcule l'aire du rectangle"""
        return self.largeur * self.hauteur
    
    def calculer_perimetre(self): # COMPORTEMENT
        """Calcule le pÃ©rimÃ¨tre du rectangle"""
        return 2 * (self.largeur + self.hauteur)
    
    def afficher(self):           # COMPORTEMENT
        """Affiche les informations du rectangle"""
        aire = self.calculer_aire()
        perimetre = self.calculer_perimetre()
        print(f"{self.nom}: {self.largeur}x{self.hauteur} - Aire: {aire}, PÃ©rimÃ¨tre: {perimetre}")
    
    def est_carre(self):          # COMPORTEMENT
        """VÃ©rifie si le rectangle est un carrÃ©"""
        return self.largeur == self.hauteur

# CrÃ©ation d'objets (instances)
rectangles = [
    Rectangle("Rect1", 10, 20),
    Rectangle("Rect2", 5, 10),
    Rectangle("Rect3", 8, 12),
    Rectangle("CarrÃ©1", 7, 7)
]

# Utilisation simple et lisible
for rect in rectangles:
    rect.afficher()
    if rect.est_carre():
        print(f"  â†’ {rect.nom} est un carrÃ© !")

# Avantages visibles :
# - Une seule liste Ã  gÃ©rer
# - DonnÃ©es et opÃ©rations regroupÃ©es
# - Code plus lisible et maintenable
# - Facile d'ajouter de nouvelles mÃ©thodes
````
